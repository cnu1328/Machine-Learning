## Update Rules (Module 02 recap)

**SGD:** $\theta \leftarrow \theta - \eta \nabla L$

**Adam:** Uses running averages $m_t$ (momentum) and $v_t$ (RMSprop) — see Module 02 Notebook 20.

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
rng = np.random.default_rng(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
model = nn.Sequential(nn.Linear(10, 5), nn.ReLU(), nn.Linear(5, 1))

# Different optimizers
optimizers = {
    'SGD': torch.optim.SGD(model.parameters(), lr=0.01),
    'SGD+Momentum': torch.optim.SGD(model.parameters(), lr=0.01, momentum=0.9),
    'Adam': torch.optim.Adam(model.parameters(), lr=0.001),
    'AdamW': torch.optim.AdamW(model.parameters(), lr=0.001, weight_decay=0.01),
}

for name, opt in optimizers.items():
    print(f'{name}: {opt}')

print('\nDay - 28: optimizer="adam"')
print('water-bodies-detection: AdamW with weight decay')
```

```python
# Training step pattern (memorize this)
def train_step(model, optimizer, X_batch, y_batch):
    optimizer.zero_grad()       # 1. Clear old gradients
    y_pred = model(X_batch)     # 2. Forward pass
    loss = F.mse_loss(y_pred, y_batch)  # 3. Compute loss
    loss.backward()             # 4. Backprop (autograd)
    optimizer.step()            # 5. Update weights
    return loss.item()

X_b = torch.randn(32, 10)
y_b = torch.randn(32, 1)
model = nn.Sequential(nn.Linear(10, 5), nn.ReLU(), nn.Linear(5, 1))
opt = torch.optim.Adam(model.parameters(), lr=0.01)
loss = train_step(model, opt, X_b, y_b)
print(f'Loss after one step: {loss:.4f}')
```

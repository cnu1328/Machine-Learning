## Batch Normalization (Ioffe & Szegedy, 2015)

Normalize each mini-batch:

$$\hat{x}_i = \frac{x_i - \mu_B}{\sqrt{\sigma_B^2 + \epsilon}}$$
$$y_i = \gamma \hat{x}_i + \beta$$

Learnable scale $\gamma$ and shift $\beta$.

**Benefits:** Faster training, higher learning rates, regularization effect.

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
rng = np.random.default_rng(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
bn = nn.BatchNorm1d(64)
x = torch.randn(32, 64) * 5 + 10  # shifted, scaled data

bn.train()
y_train = bn(x)
print(f'Before BN: mean={x.mean():.2f}, std={x.std():.2f}')
print(f'After BN (train): mean={y_train.mean():.4f}, std={y_train.std():.4f}')

bn.eval()
y_eval = bn(x)
print(f'After BN (eval):  mean={y_eval.mean():.4f}, std={y_eval.std():.4f}')
```

## Critical: model.train() vs model.eval()

- **train():** dropout ON, batch norm uses batch statistics
- **eval():** dropout OFF, batch norm uses running statistics

Forgetting `model.eval()` before inference is a common bug.

## Why Initialization Matters

- **All zeros:** all neurons learn same thing (symmetry problem)
- **Too large:** activations explode, gradients explode
- **Too small:** activations vanish, gradients vanish

**Xavier (Glorot):** $W \sim \mathcal{U}\left[-\sqrt{6/(n_{in}+n_{out})}, \sqrt{6/(n_{in}+n_{out})}\right]$ — for sigmoid/tanh

**He (Kaiming):** $W \sim \mathcal{N}(0, 2/n_{in})$ — for ReLU

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
rng = np.random.default_rng(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
def xavier_init(fan_in, fan_out):
    limit = np.sqrt(6 / (fan_in + fan_out))
    return torch.empty(fan_out, fan_in).uniform_(-limit, limit)

def he_init(fan_in, fan_out):
    return torch.randn(fan_out, fan_in) * np.sqrt(2 / fan_in)

for name, W in [('Xavier', xavier_init(784, 256)), ('He', he_init(784, 256)), ('Default', torch.randn(256, 784)*0.01)]:
    x = torch.randn(256, 784)
    out = F.relu(x @ W.T)
    print(f'{name}: output mean={out.mean():.4f}, std={out.std():.4f}')
```

```python
# PyTorch built-in
layer = nn.Linear(784, 256)
nn.init.kaiming_normal_(layer.weight, nonlinearity='relu')
print(f'Kaiming init std: {layer.weight.std():.4f}')
```

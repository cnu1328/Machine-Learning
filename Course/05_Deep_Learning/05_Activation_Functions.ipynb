## Activation Functions

| Function | Formula | Range | Use |
|----------|---------|-------|-----|
| Sigmoid | $1/(1+e^{-z})$ | (0,1) | Binary output, old hidden |
| Tanh | $(e^z-e^{-z})/(e^z+e^{-z})$ | (-1,1) | Hidden (zero-centered) |
| ReLU | $\max(0,z)$ | [0,∞) | Default hidden layer |
| Leaky ReLU | $\max(0.01z, z)$ | (-∞,∞) | Avoid dead neurons |
| GELU | $z\Phi(z)$ | (-∞,∞) | Transformers, modern CNNs |
| Softmax | $e^{z_i}/\sum e^{z_j}$ | (0,1), sum=1 | Multi-class output |

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

REPO = Path('../../').resolve()
plt.rcParams['figure.figsize'] = (8, 5)
rng = np.random.default_rng(42)
```

```python
z = np.linspace(-5, 5, 200)

activations = {
    'Sigmoid': 1 / (1 + np.exp(-z)),
    'Tanh': np.tanh(z),
    'ReLU': np.maximum(0, z),
    'Leaky ReLU': np.where(z > 0, z, 0.01 * z),
}

fig, axes = plt.subplots(1, 2, figsize=(12, 4))
for name, a in activations.items():
    axes[0].plot(z, a, label=name)
axes[0].set_title('Activation Functions'); axes[0].legend(); axes[0].grid(True, alpha=0.3)

# Derivatives
sigmoid = activations['Sigmoid']
axes[1].plot(z, sigmoid * (1 - sigmoid), label="Sigmoid'")
axes[1].plot(z, 1 - np.tanh(z)**2, label="Tanh'")
axes[1].plot(z, (z > 0).astype(float), label="ReLU'")
axes[1].set_title('Derivatives'); axes[1].legend(); axes[1].grid(True, alpha=0.3)
plt.tight_layout(); plt.show()
```

## Vanishing Gradient

Sigmoid derivative max = 0.25. Through L layers: $0.25^L \rightarrow 0$.

**ReLU** derivative = 1 for z > 0 → gradients flow without shrinking.

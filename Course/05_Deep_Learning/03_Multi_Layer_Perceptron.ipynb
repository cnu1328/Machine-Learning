## 1. Architecture

**Input layer** → **Hidden layer(s)** → **Output layer**

For layer $l$:
$$Z^{[l]} = W^{[l]} A^{[l-1]} + b^{[l]}$$
$$A^{[l]} = \sigma(Z^{[l]})$$

Where $A^{[0]} = X$ (input).

## 2. Shape Convention

$W^{[l]}$ shape: $(n^{[l]}, n^{[l-1]})$ — rows = neurons in layer l, cols = inputs from layer l-1

Day - 28: `Dense(12, input_dim=8)` → $W$ shape $(12, 8)$

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

REPO = Path('../../').resolve()
plt.rcParams['figure.figsize'] = (8, 5)
rng = np.random.default_rng(42)
```

```python
def relu(z): return np.maximum(0, z)
def sigmoid(z): return 1 / (1 + np.exp(-np.clip(z, -500, 500)))

def forward_pass(X, weights, biases, activation='relu'):
    """X: (n_samples, n_features), weights: list of (out, in) matrices"""
    A = X
    activations = [A]
    zs = []
    act_fn = relu if activation == 'relu' else sigmoid
    for W, b in zip(weights, biases):
        Z = A @ W.T + b  # (n, in) @ (in, out) + (out,) → (n, out)
        A = act_fn(Z)
        zs.append(Z)
        activations.append(A)
    return activations, zs

# Day - 28 architecture: 8 → 12 → 8 → 1
n_samples = 5
X = rng.normal(0, 1, (n_samples, 8))
W1 = rng.normal(0, 0.1, (12, 8)); b1 = np.zeros(12)
W2 = rng.normal(0, 0.1, (8, 12));  b2 = np.zeros(8)
W3 = rng.normal(0, 0.1, (1, 8));   b3 = np.zeros(1)

acts, zs = forward_pass(X, [W1, W2, W3], [b1, b2, b3], activation='relu')
for i, a in enumerate(acts):
    print(f'Layer {i} activation shape: {a.shape}')
```

## 1. Intuition

A **biological neuron** receives signals through dendrites, processes them in the cell body, and fires through the axon if the signal exceeds a threshold.

An **artificial neuron** does the same mathematically:

$$z = w_1 x_1 + w_2 x_2 + \cdots + w_n x_n + b = \mathbf{w}^T \mathbf{x} + b$$
$$a = \sigma(z)$$

- $\mathbf{x}$ = inputs (features)
- $\mathbf{w}$ = weights (learned parameters)
- $b$ = bias (shift activation threshold)
- $\sigma$ = activation function (non-linearity)

## 2. Single Neuron = Linear Regression + Activation

Without activation ($\sigma$ = identity): $\hat{y} = w^T x + b$ → **linear regression**.

With sigmoid: $\hat{y} = \sigma(w^T x + b)$ → **logistic regression**.

With ReLU: $\hat{y} = \max(0, w^T x + b)$ → building block of deep networks.

````python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

REPO = Path('../../').resolve()
plt.rcParams['figure.figsize'] = (8, 5)
rng = np.random.default_rng(42)
````

````python
def neuron(x, w, b, activation='relu'):
    z = np.dot(w, x) + b
    if activation == 'relu':
        return max(0, z), z
    elif activation == 'sigmoid':
        return 1 / (1 + np.exp(-z)), z
    return z, z

x = np.array([2.0, 3.0, 1.0])
w = np.array([0.5, -0.3, 0.8])
b = 0.1

a, z = neuron(x, w, b, 'relu')
print(f'Input: {x}')
print(f'Weights: {w}, Bias: {b}')
print(f'z = w·x + b = {z:.4f}')
print(f'a = ReLU(z) = {a:.4f}')
````

````python
# Visualize neuron response
w1_range = np.linspace(-2, 2, 100)
outputs = [neuron(np.array([w1, 1.0]), np.array([1.0, 0.5]), 0, 'relu')[0] for w1 in w1_range]
plt.plot(w1_range, outputs)
plt.xlabel('x1'); plt.ylabel('Output a')
plt.title('Single Neuron Response (ReLU)'); plt.grid(True, alpha=0.3); plt.show()
````

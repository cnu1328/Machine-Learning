## 1. Perceptron Learning Rule (Rosenblatt, 1958)

For misclassified sample $(x, y)$:

$$w_j \leftarrow w_j + \eta (y - \hat{y}) x_j$$
$$b \leftarrow b + \eta (y - \hat{y})$$

Only updates when $\hat{y} \neq y$.

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

REPO = Path('../../').resolve()
plt.rcParams['figure.figsize'] = (8, 5)
rng = np.random.default_rng(42)
```

```python
class Perceptron:
    def __init__(self, n_features, lr=0.1):
        self.w = np.zeros(n_features)
        self.b = 0.0
        self.lr = lr

    def predict(self, x):
        return 1 if np.dot(self.w, x) + self.b >= 0 else 0

    def fit(self, X, y, epochs=100):
        for _ in range(epochs):
            for xi, yi in zip(X, y):
                y_hat = self.predict(xi)
                update = self.lr * (yi - y_hat)
                self.w += update * xi
                self.b += update

# AND gate
X_and = np.array([[0,0],[0,1],[1,0],[1,1]])
y_and = np.array([0,0,0,1])

p = Perceptron(2)
p.fit(X_and, y_and)
print('AND predictions:', [p.predict(x) for x in X_and])
```

```python
# XOR fails with single perceptron
X_xor = np.array([[0,0],[0,1],[1,0],[1,1]])
y_xor = np.array([0,1,1,0])

p_xor = Perceptron(2, lr=0.1)
p_xor.fit(X_xor, y_xor, epochs=100)
print('XOR predictions (single layer):', [p_xor.predict(x) for x in X_xor])
print('XOR is NOT linearly separable → need hidden layer (MLP)')
```

```python
# Visualize decision boundary
xx, yy = np.meshgrid(np.linspace(-0.5,1.5,100), np.linspace(-0.5,1.5,100))
Z = np.array([p.predict(np.array([x,y])) for x,y in zip(xx.ravel(), yy.ravel())]).reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.3, cmap='RdYlBu')
plt.scatter(X_and[:,0], X_and[:,1], c=y_and, s=100, edgecolors='k')
plt.title('Perceptron Decision Boundary — AND gate'); plt.show()
```

## Summary

Single perceptron = linear classifier. XOR proves we need multiple layers.

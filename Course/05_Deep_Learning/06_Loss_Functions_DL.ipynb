## Loss Functions

**MSE (regression):** $L = \frac{1}{n}\sum(y - \hat{y})^2$, $\frac{\partial L}{\partial \hat{y}} = \frac{2}{n}(\hat{y} - y)$

**Binary CE:** $L = -[y\log\hat{p} + (1-y)\log(1-\hat{p})]$, with sigmoid: $\frac{\partial L}{\partial z} = \hat{p} - y$

**Categorical CE:** $L = -\sum_k y_k \log \hat{p}_k$, with softmax: $\frac{\partial L}{\partial z} = \hat{p} - y$

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
rng = np.random.default_rng(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
# PyTorch losses
y_true = torch.tensor([1.0, 0.0, 1.0])
y_pred_logits = torch.tensor([2.0, -1.0, 0.5])

bce = F.binary_cross_entropy_with_logits(y_pred_logits, y_true)
ce = F.cross_entropy(y_pred_logits.unsqueeze(0).repeat(3,1), torch.tensor([0,1,0]))
mse = F.mse_loss(torch.tensor([1.,2.,3.]), torch.tensor([1.1, 1.9, 3.2]))

print(f'BCE with logits: {bce:.4f}')
print(f'MSE: {mse:.4f}')
print('\nDay - 28 uses: loss="binary_crossentropy" → BCE + sigmoid')
```

## 1. Derivative Definition

$$f'(x) = \lim_{h \to 0} \frac{f(x+h) - f(x)}{h}$$

Geometric meaning: **slope of tangent line** at point $x$.

In ML: derivative of loss w.r.t. parameter = direction to increase loss. We move **opposite** to minimize.

## 2. Rules

| Rule | Formula |
|------|--------|
| Power | $\frac{d}{dx}x^n = nx^{n-1}$ |
| Constant | $\frac{d}{dx}c = 0$ |
| Sum | $\frac{d}{dx}[f+g] = f' + g'$ |
| Product | $\frac{d}{dx}[fg] = f'g + fg'$ |
| **Chain** | $\frac{d}{dx}f(g(x)) = f'(g(x)) \cdot g'(x)$ |

**Chain rule is the foundation of backpropagation (Module 05).**

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 11
rng = np.random.default_rng(42)
```

```python
def numerical_derivative(f, x, h=1e-7):
    return (f(x + h) - f(x - h)) / (2 * h)

f = lambda x: x**3 - 2*x**2 + x
x0 = 2.0
exact = 3*x0**2 - 4*x0 + 1  # power rule
approx = numerical_derivative(f, x0)
print(f'f(x) = x³ - 2x² + x')
print(f"f'({x0}) exact: {exact}")
print(f"f'({x0}) numerical: {approx:.6f}")
```

```python
# Visualize function and tangent
x = np.linspace(-1, 4, 200)
y = x**3 - 2*x**2 + x
x0 = 2.0; slope = 3*x0**2 - 4*x0 + 1
tangent = slope * (x - x0) + (x0**3 - 2*x0**2 + x0)

plt.plot(x, y, label='f(x)')
plt.plot(x, tangent, '--', label=f'Tangent at x={x0}, slope={slope}')
plt.axhline(0, color='k', linewidth=0.5); plt.axvline(0, color='k', linewidth=0.5)
plt.legend(); plt.title('Function and Tangent Line'); plt.show()
```

## 3. Chain Rule Example (Backprop Preview)

Let $L = (y - \hat{y})^2$ where $\hat{y} = wx + b$.

$\frac{dL}{dw} = \frac{dL}{d\hat{y}} \cdot \frac{d\hat{y}}{dw} = 2(\hat{y} - y) \cdot x$

This is the gradient for linear regression weight update.

## 1. Orthogonality

Vectors $\mathbf{u}, \mathbf{v}$ are **orthogonal** if $\mathbf{u}^T \mathbf{v} = 0$.

**Orthogonal matrix:** $Q^T Q = I$ (columns are orthonormal).

Property: $\|Q\mathbf{x}\| = \|\mathbf{x}\|$ (preserves length — used in QR decomposition).

## 2. Projection

Projection of $\mathbf{b}$ onto vector $\mathbf{a}$:
$$\text{proj}_{\mathbf{a}} \mathbf{b} = \frac{\mathbf{a}^T \mathbf{b}}{\mathbf{a}^T \mathbf{a}} \mathbf{a}$$

**ML:** Linear regression prediction = projection of $\mathbf{y}$ onto column space of $X$.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 11
rng = np.random.default_rng(42)
```

```python
a = np.array([1.0, 1.0])
b = np.array([2.0, 0.0])
proj = (a @ b) / (a @ a) * a
print(f'Projection of b onto a: {proj}')

# Residual is orthogonal to a
residual = b - proj
print(f'a · residual = {a @ residual:.2e} (should be ~0)')
```

## 3. Gram-Schmidt (Conceptual)

Orthogonalizes a set of vectors:
1. $\mathbf{u}_1 = \mathbf{v}_1$
2. $\mathbf{u}_2 = \mathbf{v}_2 - \text{proj}_{\mathbf{u}_1}\mathbf{v}_2$
3. Continue for remaining vectors

Used in QR decomposition — numerically stable alternative to normal equation.

```python
fig, ax = plt.subplots(figsize=(6, 6))
ax.quiver(0,0,*a, color='blue', scale=1, scale_units='xy', angles='xy', label='a')
ax.quiver(0,0,*b, color='green', scale=1, scale_units='xy', angles='xy', label='b')
ax.quiver(0,0,*proj, color='red', scale=1, scale_units='xy', angles='xy', label='projection')
ax.plot([proj[0], b[0]], [proj[1], b[1]], 'k--', alpha=0.5, label='residual')
ax.set_xlim(-0.5, 2.5); ax.set_ylim(-0.5, 1.5); ax.set_aspect('equal'); ax.legend(); ax.grid(True,alpha=0.3)
ax.set_title('Vector Projection'); plt.show()
```

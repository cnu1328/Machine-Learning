## 1. Likelihood

Given data $\mathbf{x} = (x_1, \ldots, x_n)$ and parameter $\theta$:
$$L(\theta) = P(\mathbf{x} | \theta) = \prod_{i=1}^{n} P(x_i | \theta)$$

**MLE:** $\hat{\theta} = \arg\max_\theta L(\theta)$

**Log-likelihood:** $\ell(\theta) = \log L(\theta) = \sum_i \log P(x_i | \theta)$ (easier to maximize).

## 2. Derivation: MLE for Gaussian Mean

For $x_i \sim \mathcal{N}(\mu, \sigma^2)$ with known $\sigma^2$:

$$\ell(\mu) = -\frac{1}{2\sigma^2}\sum_i (x_i - \mu)^2 + \text{const}$$

$$\frac{d\ell}{d\mu} = \frac{1}{\sigma^2}\sum_i (x_i - \mu) = 0$$

$$\boxed{\hat{\mu} = \frac{1}{n}\sum_{i=1}^{n} x_i}$$

**The sample mean is the MLE for Gaussian mean!**

## 3. MLE for Bernoulli → Cross-Entropy

For $y_i \in \{0, 1\}$ with $P(y=1) = p$:

$$\ell(p) = \sum_i [y_i \log p + (1-y_i)\log(1-p)]$$

Maximizing log-likelihood = **minimizing cross-entropy loss** (Module 02 Notebook 22).

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 11
rng = np.random.default_rng(42)
```

```python
data = rng.normal(5, 2, 1000)
mu_mle = data.mean()
sigma2_mle = data.var()  # MLE uses n denominator
print(f'MLE μ = {mu_mle:.3f} (sample mean)')
print(f'MLE σ² = {sigma2_mle:.3f} (sample variance)')

# Bernoulli MLE
y = rng.binomial(1, 0.7, 1000)
p_mle = y.mean()
print(f'\nMLE p (Bernoulli) = {p_mle:.3f}')
```

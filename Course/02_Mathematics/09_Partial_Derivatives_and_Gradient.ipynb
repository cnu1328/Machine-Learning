## 1. Partial Derivatives

For $f(x, y)$:
$$\frac{\partial f}{\partial x} = \lim_{h \to 0} \frac{f(x+h, y) - f(x, y)}{h}$$

Treat other variables as constants.

## 2. Gradient

$$\nabla f = \begin{bmatrix} \frac{\partial f}{\partial x_1} \\ \vdots \\ \frac{\partial f}{\partial x_n} \end{bmatrix}$$

Points in direction of **steepest ascent**. Negative gradient = steepest descent (optimization).

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 11
rng = np.random.default_rng(42)
```

```python
def f(x, y): return x**2 + 3*y**2

def grad_f(x, y):
    return np.array([2*x, 6*y])

print('∇f(1, 2) =', grad_f(1, 2))  # [2, 12]
```

## 3. Derivation: Gradient of MSE

$$L(\mathbf{w}) = \frac{1}{n}\|X\mathbf{w} - \mathbf{y}\|^2 = \frac{1}{n}(X\mathbf{w} - \mathbf{y})^T(X\mathbf{w} - \mathbf{y})$$

$$\nabla_{\mathbf{w}} L = \frac{2}{n} X^T(X\mathbf{w} - \mathbf{y})$$

At optimum: $X^T(X\mathbf{w} - \mathbf{y}) = 0$ → same as normal equation.

```python
# Verify gradient numerically
n, d = 50, 3
X = rng.normal(0, 1, (n, d))
y = rng.normal(0, 1, n)
w = rng.normal(0, 1, d)

# Analytical gradient
residual = X @ w - y
grad_analytical = (2 / n) * X.T @ residual

# Numerical gradient (per coordinate)
grad_numerical = np.zeros(d)
eps = 1e-7
for i in range(d):
    w_plus, w_minus = w.copy(), w.copy()
    w_plus[i] += eps; w_minus[i] -= eps
    L_plus = np.mean((X @ w_plus - y)**2)
    L_minus = np.mean((X @ w_minus - y)**2)
    grad_numerical[i] = (L_plus - L_minus) / (2 * eps)

print('Analytical:', grad_analytical)
print('Numerical: ', grad_numerical)
print('Max error:', np.max(np.abs(grad_analytical - grad_numerical)))
```

```python
# Visualize gradient field
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
Xg, Yg = np.meshgrid(x, y)
Z = Xg**2 + 3*Yg**2

plt.contour(Xg, Yg, Z, levels=20)
plt.quiver(Xg, Yg, 2*Xg, 6*Yg, alpha=0.6)
plt.xlabel('x'); plt.ylabel('y'); plt.title('Gradient Field of f(x,y) = x² + 3y²'); plt.show()
```

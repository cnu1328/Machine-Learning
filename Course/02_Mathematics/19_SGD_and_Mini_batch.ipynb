## 1. Three Variants

| Method | Gradient computed on | Noise | Speed |
|--------|---------------------|-------|-------|
| Batch GD | All n samples | None | Slow per epoch |
| SGD | 1 sample | High | Fast but noisy |
| Mini-batch | b samples | Moderate | Best tradeoff |

**Mini-batch SGD** is the standard in deep learning (your water-bodies `batch_size=8`).

## 2. Update Rule (Mini-batch)

$$\mathbf{w}_{t+1} = \mathbf{w}_t - \eta \nabla L_{\mathcal{B}_t}(\mathbf{w}_t)$$

where $\mathcal{B}_t$ is a random batch of size $b$.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 11
rng = np.random.default_rng(42)
```

```python
def mini_batch_gd(X, y, lr=0.1, batch_size=32, epochs=50, seed=42):
    rng = np.random.default_rng(seed)
    n, d = X.shape
    w = np.zeros(d)
    losses = []
    for epoch in range(epochs):
        idx = rng.permutation(n)
        for start in range(0, n, batch_size):
            batch = idx[start:start+batch_size]
            Xb, yb = X[batch], y[batch]
            grad = (2/len(batch)) * Xb.T @ (Xb @ w - yb)
            w -= lr * grad
        losses.append(np.mean((X @ w - y)**2))
    return w, losses

X = rng.normal(0, 1, (200, 5))
y = X @ rng.normal(0, 1, 5) + rng.normal(0, 0.1, 200)

for bs in [1, 16, 200]:
    _, losses = mini_batch_gd(X, y, batch_size=bs, epochs=30)
    plt.plot(losses, label=f'batch={bs}')
plt.xlabel('Epoch'); plt.ylabel('MSE'); plt.legend(); plt.title('Batch Size Effect'); plt.show()
```

## 1. Mean Squared Error (MSE)

$$L = \frac{1}{n}\sum_{i=1}^{n}(y_i - \hat{y}_i)^2$$

**Gradient:** $\frac{\partial L}{\partial \hat{y}_i} = \frac{2}{n}(\hat{y}_i - y_i)$

Penalizes large errors heavily (squared). Sensitive to outliers.

## 2. Mean Absolute Error (MAE)

$$L = \frac{1}{n}\sum_{i=1}^{n}|y_i - \hat{y}_i|$$

Robust to outliers. Not differentiable at zero (use subgradient).

## 3. Huber Loss

Combines MSE (small errors) and MAE (large errors):
$$L_\delta(e) = \begin{cases} \frac{1}{2}e^2 & |e| \leq \delta \\ \delta(|e| - \frac{\delta}{2}) & |e| > \delta \end{cases}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 11
rng = np.random.default_rng(42)
```

```python
errors = np.linspace(-5, 5, 200)
mse = errors**2
mae = np.abs(errors)
huber = np.where(np.abs(errors) <= 1, 0.5*errors**2, np.abs(errors) - 0.5)

plt.plot(errors, mse, label='MSE')
plt.plot(errors, mae, label='MAE')
plt.plot(errors, huber, label='Huber (δ=1)')
plt.xlabel('Error (y - ŷ)'); plt.ylabel('Loss'); plt.legend(); plt.title('Regression Loss Functions'); plt.show()
```

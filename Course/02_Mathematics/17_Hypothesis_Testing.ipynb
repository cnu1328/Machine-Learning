## 1. Hypothesis Testing Framework

1. **Null hypothesis $H_0$:** No effect (e.g., model A = model B)
2. **Alternative $H_1$:** There is an effect
3. **p-value:** Probability of observing data at least as extreme, assuming $H_0$ is true
4. If p-value < α (typically 0.05), **reject $H_0$**

## 2. Confidence Interval

95% CI for mean: $\bar{x} \pm 1.96 \frac{s}{\sqrt{n}}$

"We are 95% confident the true mean lies in this interval."

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 11
rng = np.random.default_rng(42)
```

```python
from scipy.stats import ttest_ind, ttest_rel

# Compare two models' accuracy on same test set
acc_model_a = rng.normal(0.85, 0.02, 30)
acc_model_b = rng.normal(0.87, 0.02, 30)

t_stat, p_value = ttest_rel(acc_model_a, acc_model_b)
print(f'Mean acc A: {acc_model_a.mean():.3f}')
print(f'Mean acc B: {acc_model_b.mean():.3f}')
print(f'Paired t-test p-value: {p_value:.4f}')
print(f'Significant at α=0.05?', p_value < 0.05)
```

## ML Use

- Compare two models on same test folds (paired t-test)

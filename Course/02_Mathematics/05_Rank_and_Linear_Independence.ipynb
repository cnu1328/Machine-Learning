## 1. Linear Independence

Vectors $\mathbf{v}_1, \ldots, \mathbf{v}_k$ are **linearly independent** if:
$$c_1\mathbf{v}_1 + c_2\mathbf{v}_2 + \cdots + c_k\mathbf{v}_k = \mathbf{0} \implies c_1 = c_2 = \cdots = c_k = 0$$

If one vector can be written as a combination of others → **linearly dependent** (redundant feature).

## 2. Rank

**Rank** = number of linearly independent columns (or rows) = dimension of column space.

$\text{rank}(A) \leq \min(m, n)$ for $A \in \mathbb{R}^{m \times n}$.

Full rank: $\text{rank}(A) = \min(m, n)$.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 11
rng = np.random.default_rng(42)
```

```python
A_full = np.array([[1, 0], [0, 1], [1, 1]])
A_dep = np.array([[1, 2], [2, 4], [3, 6]])  # col2 = 2*col1

print('Full rank matrix rank:', np.linalg.matrix_rank(A_full))
print('Dependent matrix rank:', np.linalg.matrix_rank(A_dep))
```

## 3. ML Connection

- **Redundant features** → rank-deficient $X^T X$ → normal equation fails
- **PCA (Module 03):** Projects to lower rank while preserving variance
- **Attention rank:** Low-rank approximations in transformers

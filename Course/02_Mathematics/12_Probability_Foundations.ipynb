## 1. Probability Axioms

1. $P(A) \geq 0$
2. $P(\Omega) = 1$ (sample space)
3. $P(A \cup B) = P(A) + P(B)$ if mutually exclusive

## 2. Conditional Probability

$$P(A|B) = \frac{P(A \cap B)}{P(B)} \quad \text{if } P(B) > 0$$

"Probability of A given B has occurred."

## 3. Bayes Theorem

$$\boxed{P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}}$$

- $P(A)$ = **prior**
- $P(B|A)$ = **likelihood**
- $P(A|B)$ = **posterior**
- $P(B)$ = **evidence**

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 11
rng = np.random.default_rng(42)
```

```python
# Medical test example
# Disease prevalence 1%, test sensitivity 99%, specificity 95%
P_D = 0.01  # prior
P_pos_given_D = 0.99  # sensitivity
P_pos_given_noD = 0.05  # false positive rate

P_pos = P_pos_given_D * P_D + P_pos_given_noD * (1 - P_D)
P_D_given_pos = P_pos_given_D * P_D / P_pos

print(f'P(disease | positive test) = {P_D_given_pos:.1%}')
print('Despite 99% sensitivity, only ~16% chance of disease!')
```

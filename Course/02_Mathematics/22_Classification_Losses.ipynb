## 1. Binary Cross-Entropy (Derive from MLE)

For $y \in \{0,1\}$, model outputs $\hat{p} = \sigma(z)$:

$$L = -[y \log \hat{p} + (1-y)\log(1-\hat{p})]$$

**Derivation from Bernoulli MLE:**
$$\ell(p) = y\log p + (1-y)\log(1-p)$$
$$L = -\ell(p) \text{ (negate to minimize)}$$

**Gradient w.r.t. logit $z$:**
$$\frac{\partial L}{\partial z} = \hat{p} - y$$

## 2. Multi-Class Cross-Entropy

With softmax $\hat{p}_k = \frac{e^{z_k}}{\sum_j e^{z_j}}$:
$$L = -\sum_{k=1}^{K} y_k \log \hat{p}_k$$

## 3. Focal Loss

$$L = -(1 - \hat{p})^\gamma \log(\hat{p})$$

Down-weights easy examples. Used in object detection (RetinaNet) and imbalanced segmentation.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 11
rng = np.random.default_rng(42)
```

```python
def binary_cross_entropy(y, p_hat, eps=1e-15):
    p_hat = np.clip(p_hat, eps, 1 - eps)
    return -(y * np.log(p_hat) + (1 - y) * np.log(1 - p_hat))

def sigmoid(z): return 1 / (1 + np.exp(-z))

# Plot BCE vs predicted probability
p = np.linspace(0.001, 0.999, 100)
plt.plot(p, binary_cross_entropy(1, p), label='y=1')
plt.plot(p, binary_cross_entropy(0, p), label='y=0')
plt.xlabel('Predicted probability'); plt.ylabel('Loss'); plt.legend(); plt.title('Binary Cross-Entropy'); plt.show()
```

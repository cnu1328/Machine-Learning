## 1. Update Rule

To minimize $L(\mathbf{w})$:
$$\mathbf{w}_{t+1} = \mathbf{w}_t - \eta \nabla L(\mathbf{w}_t)$$

- $\eta$ = **learning rate** (step size)
- $\nabla L$ = gradient at current point
- Move opposite to gradient (steepest descent)

## 2. GD for Linear Regression

$L(w) = \frac{1}{n}\sum (wx_i + b - y_i)^2$

$\frac{\partial L}{\partial w} = \frac{2}{n}\sum (wx_i + b - y_i) x_i$

$\frac{\partial L}{\partial b} = \frac{2}{n}\sum (wx_i + b - y_i)$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 11
rng = np.random.default_rng(42)
```

```python
from pathlib import Path
import pandas as pd

house = pd.read_csv(Path('../../houseprice.csv'))
X_raw = house['area'].values
y = house['price'].values
X = (X_raw - X_raw.mean()) / X_raw.std()  # normalize

w, b = 0.0, 0.0
lr = 0.1
n = len(X)
losses = []

for epoch in range(100):
    y_pred = w * X + b
    dw = (2/n) * np.sum((y_pred - y) * X)
    db = (2/n) * np.sum(y_pred - y)
    w -= lr * dw
    b -= lr * db
    loss = np.mean((y_pred - y)**2)
    losses.append(loss)

print(f'w = {w:.2f}, b = {b:.2f}')
print(f'Final MSE: {losses[-1]:,.0f}')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].plot(losses); axes[0].set_xlabel('Epoch'); axes[0].set_ylabel('MSE'); axes[0].set_title('Loss Convergence')

axes[1].scatter(X_raw, y, alpha=0.5, label='Data')
x_line = np.linspace(X_raw.min(), X_raw.max(), 100)
x_norm = (x_line - X_raw.mean()) / X_raw.std()
axes[1].plot(x_line, w * x_norm + b, 'r-', linewidth=2, label='GD fit')
axes[1].set_xlabel('Area'); axes[1].set_ylabel('Price'); axes[1].legend()
plt.tight_layout(); plt.show()
```

## 3. Learning Rate Effects

- Too large → diverges (loss increases)
- Too small → slow convergence
- Just right → smooth decrease

```python
def run_gd(lr, epochs=50):
    w, b = 0.0, 0.0
    losses = []
    for _ in range(epochs):
        y_pred = w * X + b
        w -= lr * (2/n) * np.sum((y_pred - y) * X)
        b -= lr * (2/n) * np.sum(y_pred - y)
        losses.append(np.mean((y_pred - y)**2))
    return losses

for lr in [0.001, 0.01, 0.1, 0.5]:
    plt.plot(run_gd(lr), label=f'lr={lr}')
plt.xlabel('Epoch'); plt.ylabel('MSE'); plt.legend(); plt.title('Learning Rate Comparison'); plt.yscale('log'); plt.show()
```

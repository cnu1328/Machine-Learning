## 1. Jacobian

For $\mathbf{f}: \mathbb{R}^n \to \mathbb{R}^m$:
$$J_{ij} = \frac{\partial f_i}{\partial x_j}$$

Shape: $(m \times n)$. Generalizes gradient to vector outputs.

**Backpropagation** computes Jacobians layer by layer (Module 05).

## 2. Hessian

For scalar $f: \mathbb{R}^n \to \mathbb{R}$:
$$H_{ij} = \frac{\partial^2 f}{\partial x_i \partial x_j}$$

Symmetric matrix. Describes **curvature**:
- Positive definite → convex (bowl shape, one minimum)
- Indefinite → saddle point
- Negative definite → concave

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 11
rng = np.random.default_rng(42)
```

```python
# Hessian of f(x,y) = x² + 3y²
# H = [[2, 0], [0, 6]] — positive definite (convex)
H = np.array([[2, 0], [0, 6]])
eigenvalues = np.linalg.eigvalsh(H)
print('Hessian eigenvalues:', eigenvalues)
print('Convex?', np.all(eigenvalues > 0))
```

```python
# Saddle point: f(x,y) = x² - y²
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = X**2 - Y**2

fig = plt.figure(figsize=(8, 5))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap=cm.coolwarm, alpha=0.8)
ax.set_xlabel('x'); ax.set_ylabel('y'); ax.set_zlabel('f(x,y)')
ax.set_title('Saddle Point: f(x,y) = x² - y²'); plt.show()
```

## Key Formulas

**Mean:** $\bar{x} = \frac{1}{n}\sum x_i$

**Variance:** $s^2 = \frac{1}{n-1}\sum(x_i - \bar{x})^2$

**Covariance:** $\text{Cov}(X,Y) = E[(X-E[X])(Y-E[Y])] = \frac{1}{n-1}\sum(x_i - \bar{x})(y_i - \bar{y})$

**Correlation:** $\rho = \frac{\text{Cov}(X,Y)}{\sigma_X \sigma_Y}$ (between -1 and 1)

```python
from pathlib import Path
import pandas as pd

house = pd.read_csv(Path('../../houseprice.csv'))
print(house.describe())
print('\nCorrelation matrix:')
print(house.corr())
```

```python
plt.imshow(house.corr(), cmap='RdBu_r', vmin=-1, vmax=1)
plt.colorbar(label='Correlation')
plt.xticks(range(len(house.columns)), house.columns)
plt.yticks(range(len(house.columns)), house.columns)
plt.title('Feature Correlation (House Prices)'); plt.show()
```

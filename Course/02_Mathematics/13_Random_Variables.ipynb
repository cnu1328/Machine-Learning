## 1. Random Variables

**Discrete:** Takes countable values (e.g., coin flip: 0 or 1)

**Continuous:** Takes real values (e.g., house price)

**PMF (discrete):** $P(X = x)$

**PDF (continuous):** $f(x)$ where $P(a \leq X \leq b) = \int_a^b f(x) dx$ and $\int f(x)dx = 1$

**CDF:** $F(x) = P(X \leq x)$

## 2. Expected Value and Variance

$$E[X] = \sum_x x \cdot P(X=x) \quad \text{(discrete)}$$
$$E[X] = \int x f(x) dx \quad \text{(continuous)}$$

$$\text{Var}(X) = E[(X - E[X])^2] = E[X^2] - (E[X])^2$$

$$\sigma = \sqrt{\text{Var}(X)}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 11
rng = np.random.default_rng(42)
```

```python
# Simulate coin flips
flips = rng.integers(0, 2, size=10000)
print(f'Mean (→ E[X]): {flips.mean():.3f} (expect 0.5)')
print(f'Var: {flips.var():.3f} (expect 0.25)')

# Continuous: sample from Normal
samples = rng.normal(5, 2, 10000)
print(f'\nNormal(5,4): mean={samples.mean():.2f}, std={samples.std():.2f}')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].hist(flips, bins=[-0.5,0.5,1.5], density=True, edgecolor='black')
axes[0].set_title('Discrete: Bernoulli(0.5)'); axes[0].set_xlabel('x')

axes[1].hist(samples, bins=50, density=True, alpha=0.7, label='Samples')
x = np.linspace(samples.min(), samples.max(), 200)
from scipy.stats import norm
axes[1].plot(x, norm.pdf(x, 5, 2), 'r-', linewidth=2, label='True PDF')
axes[1].set_title('Continuous: Normal(5, 4)'); axes[1].legend()
plt.tight_layout(); plt.show()
```

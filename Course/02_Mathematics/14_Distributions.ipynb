## Key Distributions

| Distribution | PMF/PDF | Parameters | ML Use |
|-------------|---------|------------|--------|
| Bernoulli | $P(X=1)=p$ | $p$ | Binary classification |
| Binomial | $\binom{n}{k}p^k(1-p)^{n-k}$ | $n, p$ | k successes in n trials |
| Gaussian | $\frac{1}{\sqrt{2\pi\sigma^2}}e^{-\frac{(x-\mu)^2}{2\sigma^2}}$ | $\mu, \sigma^2$ | Continuous features, noise |
| Poisson | $\frac{\lambda^k e^{-\lambda}}{k!}$ | $\lambda$ | Count data |

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 11
rng = np.random.default_rng(42)
```

```python
from scipy.stats import norm, bernoulli, binom, poisson

x = np.linspace(-4, 4, 200)
fig, axes = plt.subplots(2, 2, figsize=(12, 8))

axes[0,0].plot(x, norm.pdf(x, 0, 1)); axes[0,0].set_title('Gaussian N(0,1)')
axes[0,1].bar([0,1], [0.3, 0.7]); axes[0,1].set_title('Bernoulli(p=0.7)')
axes[1,0].bar(range(11), binom.pmf(range(11), 10, 0.3)); axes[1,0].set_title('Binomial(n=10, p=0.3)')
axes[1,1].bar(range(15), poisson.pmf(range(15), 3)); axes[1,1].set_title('Poisson(λ=3)')
plt.tight_layout(); plt.show()
```

## ML Connections

- **Gaussian:** Continuous features in Naive Bayes (your Day - 6), weight initialization
- **Bernoulli:** Binary cross-entropy loss (Module 02 Notebook 22)
- **Poisson:** Count targets (e.g., event counts)
- **Gaussian noise:** Assumption in linear regression

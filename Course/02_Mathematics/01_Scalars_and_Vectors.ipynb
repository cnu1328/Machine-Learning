## 1. Intuition

A **vector** is an ordered list of numbers that represents:
- A **point** in space (e.g., house features: [area, bedrooms, age])
- A **direction and magnitude** (e.g., gradient direction during training)

A **scalar** is a single number (e.g., learning rate η = 0.01, loss L = 0.45).

## 2. Mathematical Definition

**Scalar:** $c \in \mathbb{R}$ — one real number.

**Vector:** $\mathbf{v} = \begin{bmatrix} v_1 \\ v_2 \\ \vdots \\ v_n \end{bmatrix} \in \mathbb{R}^n$ — n real numbers.

**Magnitude (L2 norm):**
$$\|\mathbf{v}\| = \sqrt{v_1^2 + v_2^2 + \cdots + v_n^2} = \sqrt{\mathbf{v}^T \mathbf{v}}$$

**Unit vector:** $\hat{\mathbf{v}} = \frac{\mathbf{v}}{\|\mathbf{v}\|}$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 11
rng = np.random.default_rng(42)
```

```python
v = np.array([3.0, 4.0])
magnitude = np.linalg.norm(v)
unit = v / magnitude
print(f'v = {v}')
print(f'||v|| = {magnitude}')  # 5.0
print(f'unit vector = {unit}')
print(f'||unit|| = {np.linalg.norm(unit):.6f}')  # 1.0
```

## 3. Vector Operations

**Addition:** $\mathbf{a} + \mathbf{b} = [a_1+b_1, \ldots, a_n+b_n]$

**Scalar multiplication:** $c\mathbf{v} = [cv_1, \ldots, cv_n]$

**Dot product:** $\mathbf{a} \cdot \mathbf{b} = \sum_{i=1}^{n} a_i b_i = \mathbf{a}^T \mathbf{b}$

Geometric meaning of dot product:
$$\mathbf{a} \cdot \mathbf{b} = \|\mathbf{a}\| \|\mathbf{b}\| \cos\theta$$

If dot product = 0, vectors are **orthogonal** (perpendicular).

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
print('a + b =', a + b)
print('2 * a =', 2 * a)
print('a · b =', np.dot(a, b))  # 32

# Orthogonal vectors
x = np.array([1, 0])
y = np.array([0, 1])
print('x · y =', np.dot(x, y))  # 0
```

## 4. Visual Explanation

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Vector as arrow
ax = axes[0]
ax.quiver(0, 0, 3, 4, angles='xy', scale_units='xy', scale=1, color='blue', label='v=(3,4)')
ax.set_xlim(-1, 5); ax.set_ylim(-1, 5)
ax.set_aspect('equal'); ax.grid(True, alpha=0.3)
ax.set_title('Vector as Arrow'); ax.legend()

# Dot product and angle
ax = axes[1]
for vec, lbl, c in [([3,1],'a','blue'), ([1,2],'b','red')]: ax.quiver(0,0,*vec,color=c,angles='xy',scale_units='xy',scale=1,label=lbl)
cos_theta = np.dot([3,1],[1,2]) / (np.linalg.norm([3,1])*np.linalg.norm([1,2]))
ax.set_xlim(-1,4); ax.set_ylim(-1,3); ax.set_aspect('equal'); ax.grid(True,alpha=0.3)
ax.set_title(f'Dot product: a·b = {np.dot([3,1],[1,2])}, cos θ = {cos_theta:.3f}'); ax.legend()
plt.tight_layout(); plt.show()
```

- **Prediction:** $\hat{y} = \mathbf{w}^T \mathbf{x} + b$ (dot product!)

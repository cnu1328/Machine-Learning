## 1. Matrix Multiplication Definition

For $A \in \mathbb{R}^{m \times n}$ and $B \in \mathbb{R}^{n \times p}$:

$$(AB)_{ij} = \sum_{k=1}^{n} a_{ik} b_{kj}$$

**Shape rule:** $(m \times n)(n \times p) = (m \times p)$

Element $(i,j)$ of product = dot product of row $i$ of $A$ with column $j$ of $B$.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 11
rng = np.random.default_rng(42)
```

```python
A = np.array([[1, 2], [3, 4]])  # (2,2)
B = np.array([[5, 6], [7, 8]])  # (2,2)
C = A @ B
print('A @ B =\n', C)
# c_11 = 1*5 + 2*7 = 19
print('Manual c_11:', 1*5 + 2*7)
```

## 2. Properties

1. **Associative:** $(AB)C = A(BC)$
2. **NOT commutative:** $AB \neq BA$ in general
3. **Identity:** $AI = IA = A$
4. **Transpose:** $(AB)^T = B^T A^T$

```python
A = np.array([[1, 2], [3, 4]])
B = np.array([[0, 1], [1, 0]])
print('AB =\n', A @ B)
print('BA =\n', B @ A)
print('Equal?', np.allclose(A @ B, B @ A))
```

## 3. Derivation: Normal Equation for Linear Regression

**Problem:** Given $X \in \mathbb{R}^{n \times d}$ and $\mathbf{y} \in \mathbb{R}^n$, find $\mathbf{w}$ minimizing:
$$L(\mathbf{w}) = \|X\mathbf{w} - \mathbf{y}\|^2 = (X\mathbf{w} - \mathbf{y})^T(X\mathbf{w} - \mathbf{y})$$

**Step 1:** Expand:
$$L = \mathbf{w}^T X^T X \mathbf{w} - 2\mathbf{y}^T X \mathbf{w} + \mathbf{y}^T \mathbf{y}$$

**Step 2:** Take gradient (Module 02 Notebook 09):
$$\nabla_{\mathbf{w}} L = 2X^T X \mathbf{w} - 2X^T \mathbf{y}$$

**Step 3:** Set to zero:
$$X^T X \mathbf{w} = X^T \mathbf{y}$$

**Step 4:** Normal equation:
$$\boxed{\hat{\mathbf{w}} = (X^T X)^{-1} X^T \mathbf{y}}$$

```python
# Verify normal equation on house price data
from pathlib import Path
import pandas as pd

house = pd.read_csv(Path('../../houseprice.csv'))
X = house[['area']].values
y = house['price'].values

# Add bias column of ones
X_b = np.column_stack([np.ones(len(X)), X])

# Normal equation
w_hat = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
print(f'Intercept: {w_hat[0]:,.2f}')
print(f'Slope (area): {w_hat[1]:.2f}')

# Compare with sklearn
from sklearn.linear_model import LinearRegression
m = LinearRegression().fit(X, y)
print(f'sklearn intercept: {m.intercept_:,.2f}, slope: {m.coef_[0]:.2f}')
```

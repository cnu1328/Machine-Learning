## 1. Convexity

$f$ is **convex** if for all $x, y$ and $\lambda \in [0,1]$:
$$f(\lambda x + (1-\lambda)y) \leq \lambda f(x) + (1-\lambda)f(y)$$

Convex → any local minimum is global minimum. Linear regression MSE is convex.

## 2. Local vs Global Minima

Neural networks have **non-convex** loss landscapes with many local minima and saddle points. In practice, SGD finds good enough solutions.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 11
rng = np.random.default_rng(42)
```

```python
# Non-convex function
x = np.linspace(-3, 3, 500)
f = lambda x: x**4 - 4*x**2 + x
plt.plot(x, f(x)); plt.axhline(0, color='k', linewidth=0.5)
plt.xlabel('x'); plt.ylabel('f(x)'); plt.title('Non-convex: multiple local minima'); plt.show()
```

```python
# 2D loss landscape
w1 = np.linspace(-2, 2, 100)
w2 = np.linspace(-2, 2, 100)
W1, W2 = np.meshgrid(w1, w2)
L = W1**2 + 5*W2**2  # convex bowl

plt.contour(W1, W2, L, levels=30)
plt.xlabel('w1'); plt.ylabel('w2'); plt.title('Convex Loss Landscape (MSE-like)'); plt.show()
```

## ML Insight

- **Linear/logistic regression:** Convex → guaranteed global optimum
- **Neural networks:** Non-convex → need good initialization, learning rate, and luck
- **Saddle points** are more common than local minima in high dimensions

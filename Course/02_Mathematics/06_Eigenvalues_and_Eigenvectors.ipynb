## 1. Definition

For square matrix $A \in \mathbb{R}^{n \times n}$, if:
$$A\mathbf{v} = \lambda \mathbf{v}$$

for non-zero vector $\mathbf{v}$ and scalar $\lambda$, then:
- $\lambda$ = **eigenvalue**
- $\mathbf{v}$ = **eigenvector**

Intuition: eigenvectors are directions unchanged by $A$ (only scaled by $\lambda$).

## 2. Characteristic Equation

$\det(A - \lambda I) = 0$ → polynomial whose roots are eigenvalues.

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 11
rng = np.random.default_rng(42)
```

```python
A = np.array([[4, 1], [2, 3]])
eigenvalues, eigenvectors = np.linalg.eig(A)
print('Eigenvalues:', eigenvalues)
print('Eigenvectors (columns):\n', eigenvectors)

# Verify Av = λv
for i in range(len(eigenvalues)):
    lam, v = eigenvalues[i], eigenvectors[:, i]
    Av = A @ v
    lv = lam * v
    print(f'λ={lam:.2f}: ||Av - λv|| = {np.linalg.norm(Av - lv):.2e}')
```

## 3. Derivation: PCA from Eigendecomposition

Given centered data matrix $X_c$ (n × d):

1. Compute **covariance matrix:** $C = \frac{1}{n} X_c^T X_c$ (symmetric d × d)
2. Eigendecompose: $C = V \Lambda V^T$
3. Eigenvectors $V$ = principal directions
4. Eigenvalues $\Lambda$ = variance along each direction
5. Project: $X_{\text{PCA}} = X_c V_k$ (keep top k eigenvectors)

This is exactly your `Day - 21 Plant Iris Clustering Using PCA`.

```python
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data
X_c = X - X.mean(axis=0)

# Covariance and eigendecomposition
C = (X_c.T @ X_c) / len(X_c)
eigenvalues, eigenvectors = np.linalg.eigh(C)

# Sort descending
idx = np.argsort(eigenvalues)[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]

# Project to 2D
X_pca = X_c @ eigenvectors[:, :2]

fig, ax = plt.subplots()
for label, color in zip([0,1,2], ['r','g','b']):
    mask = iris.target == label
    ax.scatter(X_pca[mask, 0], X_pca[mask, 1], c=color, label=iris.target_names[label], alpha=0.7)
ax.set_xlabel(f'PC1 ({eigenvalues[0]/eigenvalues.sum():.1%} var)')
ax.set_ylabel(f'PC2 ({eigenvalues[1]/eigenvalues.sum():.1%} var)')
ax.legend(); ax.set_title('PCA via Eigendecomposition (Iris)'); plt.show()
```

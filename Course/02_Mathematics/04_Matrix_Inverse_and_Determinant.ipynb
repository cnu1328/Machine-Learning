## 1. Determinant (2×2)

For $A = \begin{bmatrix} a & b \\ c & d \end{bmatrix}$:
$$\det(A) = ad - bc$$

Geometric meaning: absolute value = area scaling factor of the linear transformation.

$\det(A) = 0$ → matrix is **singular** (not invertible).

## 2. Matrix Inverse

$A^{-1}$ satisfies: $AA^{-1} = A^{-1}A = I$

Exists **if and only if** $\det(A) \neq 0$.

**2×2 formula:**
$$A^{-1} = \frac{1}{ad-bc} \begin{bmatrix} d & -b \\ -c & a \end{bmatrix}$$

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

plt.rcParams['figure.figsize'] = (8, 5)
plt.rcParams['font.size'] = 11
rng = np.random.default_rng(42)
```

```python
A = np.array([[4.0, 7.0], [2.0, 6.0]])
det = np.linalg.det(A)
A_inv = np.linalg.inv(A)
print(f'det(A) = {det}')
print(f'A @ A_inv =\n{A @ A_inv}')

# Solve Ax = b
b = np.array([1, 2])
x = A_inv @ b
print(f'Solution x = {x}')
print(f'Verify Ax = {A @ x}')
```

## 3. Why Normal Equation Can Fail

If $X^T X$ is singular (columns of $X$ are linearly dependent), $(X^T X)^{-1}$ does not exist.

**Solution in practice:** Use gradient descent or add regularization (Ridge): $(X^T X + \lambda I)^{-1} X^T y$

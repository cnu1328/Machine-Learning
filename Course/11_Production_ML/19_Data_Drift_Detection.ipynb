## Data Drift

**Covariate drift:** Input distribution changes (new satellite sensor, season, region).

**Label drift:** P(y) changes (new pond types, drought).

**Detection methods:**
- PSI (Population Stability Index) on band statistics
- KS test on NDVI/NDWI distributions
- Embedding distance (DINO features)

**GeoSpatial triggers:** New Planet collection, different sun angle, new geography.

```python
import os
import json
import time
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
def psi(expected, actual, bins=10):
    """Population Stability Index — >0.25 often flagged."""
    e_hist, _ = np.histogram(expected, bins=bins, density=True)
    a_hist, _ = np.histogram(actual, bins=bins, density=True)
    e_hist = np.clip(e_hist, 1e-6, None)
    a_hist = np.clip(a_hist, 1e-6, None)
    return float(np.sum((a_hist - e_hist) * np.log(a_hist / e_hist)))

train_bands = np.random.normal(0.3, 0.1, 1000)
new_bands = np.random.normal(0.45, 0.15, 500)  # shifted distribution
print(f'PSI (shifted): {psi(train_bands, new_bands):.3f}')
```

## Why Track Experiments?

Without tracking: "Which run had IoU 0.82? What config? What weights?"

**Log per run:**
- Hyperparameters (full YAML)
- Metrics per epoch (loss, IoU, Dice)
- Artifacts (best.pt, config, plots)
- Environment (git commit, package versions)

````python
import os
import json
import time
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
````

````python
# MLflow pattern (concept)
try:
    import mlflow
    mlflow.set_experiment('water-bodies')
    with mlflow.start_run(run_name='unetpp-se_resnet50-v1'):
        mlflow.log_params({'lr': 2e-4, 'batch_size': 2, 'encoder': 'se_resnet50'})
        for epoch in range(3):
            mlflow.log_metrics({'train_loss': 0.5 - epoch*0.1, 'val_iou': 0.6 + epoch*0.05}, step=epoch)
        print('MLflow run logged')
except ImportError:
    print('Optional: pip install mlflow')

# W&B: wandb.init(project='water-bodies', config=cfg)
````

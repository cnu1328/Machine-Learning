## Pruning

**Unstructured:** Zero individual weights — needs sparse hardware support.

**Structured:** Remove entire channels/filters — direct speedup on standard GPUs.

**Workflow:** Train → prune → fine-tune → export.

```python
import os
import json
import time
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
import torch.nn.utils.prune as prune
layer = nn.Conv2d(32, 64, 3, padding=1)
prune.l1_unstructured(layer, name='weight', amount=0.3)
sparsity = (layer.weight == 0).float().mean().item()
print(f'Sparsity after 30% prune: {sparsity:.1%}')
```

Pruning reduces model size; structured pruning gives real speedups.

## DVC (Data Version Control)

**Problem:** Git cannot store 50GB of GeoTIFF tiles.

**DVC solution:**
- Store data in S3/GCS/local cache
- Git tracks `.dvc` pointer files + `dvc.yaml` pipelines
- `dvc repro` rebuilds pipeline from raw data

```
raw/ → dvc stage tile → tiles/ → dvc stage train → models/
```

````python
import os
import json
import time
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
````

````python
# dvc.yaml pipeline (conceptual)
pipeline = '''
stages:
  tile:
    cmd: python tile_and_mask.py --config config/default.yaml
    deps:
      - tile_and_mask.py
      - config/default.yaml
    outs:
      - tiles/
      - masks_aqua/
      - masks_boundary/
  train:
    cmd: python train.py --config config/default.yaml
    deps:
      - train.py
      - tiles/
    outs:
      - runs/latest/
'''
print(pipeline)
````

## GeoSpatial Data Versioning

Version together:
- Raw GeoTIFF + shapefile (DVC remote)
- `config/default.yaml` (git)
- Tile output hash (DVC outs)
- Model weights (MLflow artifact or DVC)

**Rule:** Never train on unversioned data.

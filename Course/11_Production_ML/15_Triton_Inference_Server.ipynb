## Triton Inference Server

**Features:**
- Multi-model serving (PyTorch, ONNX, TensorRT)
- Dynamic batching — groups requests for GPU efficiency
- Model ensemble pipelines
- Metrics endpoint for Prometheus

**When to use:** High-throughput GPU serving, multiple models, SLA requirements.

**When FastAPI suffices:** Low QPS, simple single-model API.

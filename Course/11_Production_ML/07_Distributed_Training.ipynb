## Distributed Data Parallel (DDP)

Each GPU:
- Gets a full model copy
- Processes different batch shard
- Gradients all-reduced across GPUs

**Launch:**
```bash
torchrun --nproc_per_node=2 train.py --config config/default.yaml
```

**When worth it:** Large dataset, batch size limited by GPU memory.

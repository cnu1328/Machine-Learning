## Docker for ML

**Why containers:** Same environment dev → staging → production.

**water-bodies-detection Dockerfile:**
```dockerfile
FROM python:3.11-slim
RUN apt-get install gdal-bin libgdal-dev libgeos-dev ...
COPY requirements.txt .
RUN pip install -r requirements.txt
COPY . /app
CMD ["python", "predict.py", "--help"]
```

**GeoSpatial deps:** GDAL, PROJ, GEOS — must match host or run fully containerized.

## Best Practices

- Pin Python + package versions in requirements.txt
- Multi-stage builds for smaller images
- Mount data volumes, not COPY large GeoTIFFs
- GPU: `nvidia/cuda` base + `--gpus all`

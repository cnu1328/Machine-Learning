## Model Registry

**Track per version:**
- Weights file (best.pt)
- Training config (frozen YAML)
- Metrics (val IoU, Dice)
- Data hash / DVC revision
- Git commit SHA
- Deployment status (staging/prod)

**water-bodies:** `model_meta.json` in each run folder documents architecture, bands, thresholds.

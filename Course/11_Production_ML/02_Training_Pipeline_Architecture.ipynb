## Config-Driven Design

**Single source of truth:** All hyperparameters in YAML, not hardcoded.

**Benefits:**
- Reproduce any experiment from config file
- Git-diff config changes between runs
- Non-engineers can tune without code changes

## water-bodies config/default.yaml Sections

| Section | Key params |
|---------|------------|
| `data` | input_size, band_indices, batch_size |
| `tiling` | tile_size, overlap, boundary_width_meters |
| `model` | unetplusplus, se_resnet50, out_channels=2 |
| `training` | two-stage LR, early stopping, loss weights |
| `inference` | thresholds, TTA, sliding window |

## ML CI/CD Pipeline

```
Push code → CI:
  - lint, unit tests
  - smoke train (1 epoch)
  - inference test on fixture tile
  - compare IoU vs baseline

Merge → CD (if metrics pass):
  - build Docker image
  - push to registry
  - deploy to staging
  - run integration tests
  - promote to prod
```

**GitHub Actions / GitLab CI** — same as software, plus model metric gates.

````python
import os
import json
import time
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
````

````python
ci_pipeline = '''
name: water-bodies-ci
on: [push, pull_request]
jobs:
  test:
    runs-on: ubuntu-latest
    steps:
      - uses: actions/checkout@v4
      - run: pip install -r requirements.txt
      - run: pytest tests/
      - run: python predict.py --help
  smoke-train:
    runs-on: [self-hosted, gpu]
    steps:
      - run: python train.py --epochs 1 --config config/smoke.yaml
'''
print(ci_pipeline)
````

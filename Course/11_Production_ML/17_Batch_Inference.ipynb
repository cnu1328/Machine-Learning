## Batch Inference

**water-bodies automate/automate_water_predictions.py:**
```
For each GeoTIFF in input_dir:
  → subprocess predict.py
  → write stem_aqua.tif, stem_boundary.tif
  → log success/failure
```

**Patterns:**
- Idempotent outputs (skip if exists)
- Parallel workers (careful with GPU memory)
- Checkpoint/resume on failure
- Structured logs per file

````python
import os
import json
import time
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
````

````python
# Batch job tracker (concept)
files = ['tile_001.tif', 'tile_002.tif', 'tile_003.tif']
status = {}
for f in files:
    try:
        # predict(f)
        status[f] = 'ok'
    except Exception as e:
        status[f] = f'fail: {e}'
ok = sum(1 for v in status.values() if v == 'ok')
print(f'Batch complete: {ok}/{len(files)} succeeded')
````

```python
import os
import json
import time
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
# Weekly QA sample tracker
qa_results = [{'week': 1, 'sample_iou': 0.86}, {'week': 2, 'sample_iou': 0.84}, {'week': 3, 'sample_iou': 0.79}]
weeks = [r['week'] for r in qa_results]
ious = [r['sample_iou'] for r in qa_results]
plt.plot(weeks, ious, 'o-'); plt.axhline(0.82, color='r', ls='--', label='Alert threshold')
plt.xlabel('Week'); plt.ylabel('Sample IoU'); plt.legend(); plt.title('Production QA trend'); plt.show()
```

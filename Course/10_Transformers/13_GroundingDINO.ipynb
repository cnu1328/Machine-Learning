## When to Use Transformers in Production

| Use Case | Recommendation |
|----------|---------------|
| Aquaculture pond boundaries | UNet++ (CNN) — proven |
| Rapid prototype new class | SAM + CLIP prompts |
| Unlabeled pretraining | DINO on satellite tiles |
| Multi-class land cover | SegFormer |
| Open-vocab exploration | Grounding DINO |
| Real-time detection | YOLOv8 (not transformer) |

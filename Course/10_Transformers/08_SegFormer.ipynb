## SegFormer (Xie et al., 2021)

**Efficient semantic segmentation transformer:**

- **Encoder:** Hierarchical MiT (Mix Transformer) — like Swin-lite
- **Decoder:** Lightweight all-MLP (no heavy upsampling conv)
- **No positional encoding** in decoder — saves compute

**Performance:** Competitive mIoU on ADE20K/Cityscapes with fewer params than many CNN seg models.

## SegFormer vs Your UNet++

| | UNet++ (water-bodies) | SegFormer |
|---|----------------------|----------|
| Backbone | SE-ResNet50 (CNN) | MiT (Transformer) |
| Inductive bias | Strong spatial | Weak, data-driven |
| Multi-band input | 6 Planet bands | Flexible |
| Production maturity | Your deployed pipeline | Research/modern alt |
| Best when | Proven GeoSpatial workflow | Large labeled land cover |

**Recommendation:** Keep UNet++ for aquaculture; evaluate SegFormer for multi-class land cover at scale.

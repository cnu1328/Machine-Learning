## DINO (Caron et al., 2021)

**Self-supervised learning** for ViT without labels.

**Method:**
1. Student network sees augmented crop
2. Teacher network (EMA of student) sees different crop
3. Match teacher output distribution to student
4. Teacher provides stable targets

**Result:** ViT features cluster semantically without labels — excellent for downstream segmentation with few labels.

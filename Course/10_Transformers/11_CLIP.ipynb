## CLIP (Radford et al., 2021)

**Contrastive Language-Image Pretraining:**

- Image encoder (ViT or ResNet) + text encoder (Transformer)
- Train on 400M image-text pairs
- Loss: maximize cosine similarity of matching pairs, minimize non-matching

**Zero-shot:** Classify by comparing image embedding to text prompts:
- "a satellite image of aquaculture pond"
- "a satellite image of bare soil"
- "a satellite image of vegetation"

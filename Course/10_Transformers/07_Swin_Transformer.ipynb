## Swin Transformer (Liu et al., 2021)

**Problems with ViT:** Quadratic cost on patch count; fixed resolution; no hierarchy.

**Swin solutions:**
1. **Window attention:** Self-attention within local $M \times M$ windows → $O(n)$ per layer
2. **Shifted windows:** Alternate layers shift window partition for cross-window connections
3. **Hierarchical:** Patch merging between stages (like CNN pyramid)

**Result:** ViT-like accuracy with CNN-like efficiency for dense prediction.

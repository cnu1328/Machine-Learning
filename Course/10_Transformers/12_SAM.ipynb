## SAM (Kirillov et al., 2023)

**Promptable segmentation foundation model:**

1. **Image encoder:** ViT-H (heavy, run once per image)
2. **Prompt encoder:** Points, boxes, or coarse masks
3. **Mask decoder:** Lightweight transformer → 3 mask candidates + IoU scores

**Training:** 1.1B masks on 11M images — segmentation foundation model.

**Zero-shot:** Works on new domains without fine-tuning (with varying quality).

## Training Output Directory

```
models/run_YYYYMMDD_HHMMSS/
├── best.pt              # Highest val_iou_aqua
├── final.pt             # Last epoch weights
├── model_meta.json      # Architecture metadata
└── train_config.yaml    # Frozen config copy
```

## model_meta.json

```json
{
  "in_channels": 6,
  "out_channels": 2,
  "input_size": 512,
  "best_val_iou_aqua": 0.87,
  "tiles_dir": "...",
  "masks_aqua_dir": "...",
  "masks_boundary_dir": "..."
}
```

**predict.py reads this** to rebuild model with correct channels.

## Resume

```bash
python train.py ... --resume path/to/best.pt
```

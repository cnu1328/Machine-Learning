## The Core Problem

```
[Pond A][bund][Pond B]  ← single semantic head → one water blob
[Pond A] | [Pond B]     ← dual head → bund separates → 2 polygons
```

## Three Solutions Compared

| Approach | Labels | Inference | Your Project |
|----------|--------|-----------|-------------|
| Single-head aqua | Interior only | CC splits (fragile) | ❌ |
| **Dual-head aqua+boundary** | Interior + bund ring | `aqua & ~boundary` | ✅ Production |
| Mask R-CNN instance | Instance polygons | Native IDs | Alternative (Mod 09) |

## Why boundary weight = 0.35?

Boundary pixels are **sparse** (~1–3 px wide). Without down-weighting, boundary gradients would dominate and destabilize aqua learning.

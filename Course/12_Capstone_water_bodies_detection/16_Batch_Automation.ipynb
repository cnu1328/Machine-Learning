## automate/automate_water_predictions.py

```python
for each GeoTIFF in input_dir:
    subprocess.run([python, predict.py, ...])
    → writes stem_aqua.tif, stem_boundary.tif
```

**Features:**
- Streams subprocess stdout to console
- `--model_dir` resolves weights + model_meta.json
- Idempotent if outputs exist (check script logic)

## automate/automate_water_postprocess.py

Batch wrapper for `post_process_aqua_boundary.py` over prediction pairs.

## Production Pattern

```bash
# Step 1: Batch predict
python automate/automate_water_predictions.py \
  --model_dir models/run_20260115/ \
  --input_dir /data/mosaics/ \
  --output_dir /data/predictions/

# Step 2: Batch post-process
python automate/automate_water_postprocess.py \
  --predictions_dir /data/predictions/ \
  --output_dir /data/shapefiles/
```

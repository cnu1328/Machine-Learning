## Post-Process Pipeline

```
1. Gaussian smooth probability maps (sigma=0.3)
2. Threshold: aqua >= 0.8, boundary >= 0.5
3. Morphology: close boundary gaps (boundary_close_px=1)
4. interior = aqua_mask & ~boundary_mask
5. fill_holes(interior)
6. connected_components (connectivity=8)
7. filter min_pond_area (300 m²)
8. vectorize → shapely polygons
9. smooth (buffer +/-) + simplify (Douglas-Peucker)
10. write aqua_ponds.shp
```

## Key Parameters

| Param | Default | Fixes |
|-------|---------|-------|
| `threshold_aqua` | 0.8 | False positives |
| `boundary_close_px` | 1 | Gaps in bund lines |
| `min_pond_area` | 300 m² | Tiny noise blobs |
| `smooth_distance` | 2.0 | Jagged raster edges |
| `simplify_tolerance` | 1.5 | Vertex count |
| `fill_holes` | true | Donut artifacts |

```bash

# 1. Tile + masks
python tile_and_mask.py \
  --input_tif area.tif --input_shp ponds.shp \
  --output_dir tiles_masks/ --config config/default.yaml

# 2. Train
python train.py \
  --tiles_dir tiles_masks/tiles \
  --masks_aqua_dir tiles_masks/masks_aqua \
  --masks_boundary_dir tiles_masks/masks_boundary \
  --config config/default.yaml --output_dir models/

# 3. Predict
python predict.py \
  --input_tif mosaic.tif --output_dir preds/ \
  --model_dir models/run_YYYYMMDD/ --tta d4

# 4. Post-process
python post_process/post_process_aqua_boundary.py \
  --config config/default.yaml \
  --aqua_tif preds/mosaic_aqua.tif \
  --boundary_tif preds/mosaic_boundary.tif \
  --output_path shp/
```

## Debug Guide

| Symptom | Likely Cause | Fix |
|---------|-------------|-----|
| IoU stuck at 0 | Wrong mask paths, empty masks | Check triplet filenames |
| NaN loss | Bad normalization, all-nodata tiles | Check valid pixel fraction |
| Merged ponds in output | Boundary head weak | More boundary labels, tune loss weight |
| Seam artifacts | Overlap too low | Increase overlap_fraction |
| CRS errors | SHP vs TIF mismatch | Reproject in tile_and_mask |
| OOM on GPU | batch_size too large | Reduce to 1, enable AMP |

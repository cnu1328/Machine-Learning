## Dockerfile

```dockerfile
FROM python:3.11-slim
RUN apt-get install gdal-bin libgdal-dev libgeos-dev ...
COPY requirements.txt .
RUN pip install -r requirements.txt
COPY . /app
CMD ["python", "predict.py", "--help"]
```

**Why GDAL in apt?** rasterio/geopandas need native libs — pip alone insufficient.

## Run

```bash
docker build -t water-bodies-detection .
docker run --gpus all --shm-size=16g \
  -v /data:/data \
  water-bodies-detection \
  python predict.py --input_tif /data/mosaic.tif ...
```

**--shm-size=16g:** PyTorch DataLoader shared memory for multi-worker loading.

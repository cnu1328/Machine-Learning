## Learning Objectives

3. Preserve georeferencing in outputs

## Output Files

| File | Content | Dtype |
|------|---------|-------|
| `*_aqua.tif` | P(pond interior) per pixel | float32 [0,1] |
| `*_boundary.tif` | P(bund) per pixel | float32 [0,1] |

**Georeferencing:** Same CRS + transform as input mosaic (via rasterio profile).

## Threshold Strategy

| Stage | aqua | boundary | Rationale |
|-------|------|----------|----------|
| Inference default | 0.5 | 0.5 | Balanced probability maps |
| Post-process | **0.8** | 0.5 | High-precision GIS polygons |

**Key insight:** Keep inference recall-friendly; tighten at post-process for GIS delivery.

## Problem Statement

**Task:** Detect and delineate **aquaculture ponds** from Planet multispectral imagery.

**Not simple water detection:** Adjacent ponds share walls (bunds). Single-class semantic segmentation merges them into one blob.

**Solution:** Dual-head UNet++ predicts:
- **ch0 — Aqua interior:** filled pond polygon
- **ch1 — Bund boundary:** earthen divider lines

Post-process: `interior = aqua & ~boundary` → connected components → one polygon per pond.

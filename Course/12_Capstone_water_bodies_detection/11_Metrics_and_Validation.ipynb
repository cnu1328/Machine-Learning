## Metric Definitions

| Metric | Formula | Use |
|--------|---------|-----|
| IoU | \|P∩T\|/\|P∪T\| | Checkpoint selection |
| Dice | 2\|P∩T\|/(\|P\|+\|T\|) | Loss component |
| Precision | TP/(TP+FP) | False pond rate |
| Recall | TP/(TP+FN) | Missed ponds |

## Validation Caveat

Tile-level split can **leak spatial correlation** (tiles from same mosaic in train and val). For rigorous benchmarks, use raster-level splits.

```python
def build_water_model(in_channels, out_channels=2, ...):
    return smp.UnetPlusPlus(
        encoder_name='se_resnet50',
        encoder_weights='imagenet',
        in_channels=6,
        classes=2,
        activation=None,  # raw logits
        decoder_attention_type='scse',
        decoder_channels=(256, 128, 64, 32, 16),
    )

def set_encoder_trainable(model, trainable: bool):
    for p in model.encoder.parameters():
        p.requires_grad = trainable
```

## Architecture Choices

| Component | Choice | Rationale |
|-----------|--------|----------|
| UNet++ | Nested dense skips | Best edge quality for water/ponds |
| SE-ResNet50 | ImageNet pretrain + SE blocks | Channel recalibration for 6-band input |
| SCSE decoder | Spatial + channel attention | Focus on thin bund pixels |
| activation=None | Raw logits | BCEWithLogitsLoss in losses.py |
| out_channels=2 | Aqua + boundary | Dual-head design |

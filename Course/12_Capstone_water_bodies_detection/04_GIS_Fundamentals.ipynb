## Core GIS Concepts

| Concept | Meaning | In Project |
|---------|---------|------------|
| **CRS** | Coordinate reference system | Shapefile reprojected to match raster |
| **Transform** | Pixel → map coordinates | Preserved in output GeoTIFFs |
| **GSD** | Ground sample distance (m/px) | `meters_per_pixel` for bund width |
| **Rasterize** | Vector → pixel mask | Aqua + boundary mask generation |
| **Vectorize** | Pixel labels → polygons | post_process output .shp |

## Libraries

- **rasterio:** GeoTIFF I/O, windows, rasterize, shapes
- **geopandas:** Shapefile reading, CRS ops
- **shapely:** Geometry validation, buffer, simplify

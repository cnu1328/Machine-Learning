## Augmentation (`build_train_augmentation`)

**Critical:** `additional_targets={'mask_boundary': 'mask'}` — both masks transform together.

| Aug | p | Why |
|-----|---|-----|
| H/V flip | 0.5 | No preferred orientation |
| ShiftScaleRotate | 0.55 | GSD/view variation |
| ElasticTransform/GridDistortion | 0.25 | Irregular pond shapes |
| GaussNoise/Blur | 0.35 | Sensor/atmosphere |
| RandomGamma | 0.45 | Sun angle variation |

**Validation:** `augment=False` — no random transforms.

```python
# __getitem__ flow (conceptual)
# 1. rasterio.read tile → (C,H,W)
# 2. load masks → (H,W) each
# 3. if augment: albumentations on image + both masks
# 4. stack masks → (2,H,W) tensor
print('Output: image (6,512,512), mask (2,512,512)')
```

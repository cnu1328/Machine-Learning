## 1. Intuition

**Reinforcement Learning (RL):** An **agent** takes **actions** in an **environment** to maximize **cumulative reward**.

Unlike supervised learning:
- No labeled (input, output) pairs
- Agent discovers good actions through trial and error
- Delayed rewards (action now affects future outcomes)

## 2. Markov Decision Process (MDP)

$(S, A, P, R, \gamma)$:
- $S$ = states
- $A$ = actions
- $P(s'|s,a)$ = transition probability
- $R(s,a)$ = reward
- $\gamma$ = discount factor

**Goal:** Find policy $\pi(a|s)$ maximizing $G = \sum_{t=0}^{\infty} \gamma^t R_t$

## 3. Multi-Armed Bandit & UCB

Simplest RL problem: $K$ slot machines (arms), each with unknown reward rate $\mu_k$.

**Exploration vs Exploitation:**
- Exploit: pull best-known arm
- Explore: try other arms to discover better ones

**Upper Confidence Bound (UCB):**
$$a_t = \arg\max_k \left[ \hat{\mu}_k + \sqrt{\frac{2\ln t}{n_k}} \right]$$

Your **Day - 25** implements UCB for web ad optimization.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

REPO = Path('../../').resolve()
plt.rcParams['figure.figsize'] = (8, 5)
rng = np.random.default_rng(42)
```

```python
# Multi-Armed Bandit simulation
n_arms = 5
true_means = rng.uniform(0.2, 0.8, n_arms)
print('True arm means:', np.round(true_means, 3))

# UCB algorithm
counts = np.zeros(n_arms)
values = np.zeros(n_arms)
total_reward = 0
rewards_history = []

for t in range(1, 5001):
    if t <= n_arms:
        arm = t - 1  # explore each arm once
    else:
        ucb = values + np.sqrt(2 * np.log(t) / np.maximum(counts, 1))
        arm = ucb.argmax()
    reward = rng.random() < true_means[arm]
    counts[arm] += 1
    values[arm] += (reward - values[arm]) / counts[arm]
    total_reward += reward
    rewards_history.append(reward)

print(f'Best arm: {true_means.argmax()}, Selected most: {counts.argmax()}')
print(f'Arm pull counts: {counts.astype(int)}')
```

```python
plt.plot(pd.Series(rewards_history).rolling(100).mean())
plt.xlabel('Round'); plt.ylabel('Avg reward (100-round window)')
plt.title('UCB Multi-Armed Bandit — Day - 25 Concept'); plt.show()
```

## 4. Q-Learning (Preview)

$$Q(s,a) \leftarrow Q(s,a) + \alpha [R + \gamma \max_{a'} Q(s',a') - Q(s,a)]$$

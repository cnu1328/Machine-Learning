## 1. Intuition

**Unsupervised learning** = find structure in data **without labels**.

Two main goals:
1. **Clustering:** group similar samples (customer segments, land cover types)
2. **Dimensionality reduction:** compress features while preserving information (PCA, visualization)

## 2. Mathematical Framework

No target $y$. Optimize objectives like:

- **K-Means:** $\min \sum_k \sum_{x_i \in C_k} \|x_i - \mu_k\|^2$
- **PCA:** $\max_v v^T C v$ subject to $\|v\|=1$ (maximize variance)
- **Autoencoder:** $\min \|x - \text{decode}(\text{encode}(x))\|^2$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

REPO = Path('../../').resolve()
plt.rcParams['figure.figsize'] = (8, 5)
rng = np.random.default_rng(42)
```

```python
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris

X, y_true = load_iris(return_X_y=True)  # y_true only for visualization

# Clustering (unsupervised — don't use y_true for training)
km = KMeans(n_clusters=3, random_state=42, n_init=10).fit(X)
print('K-Means cluster sizes:', np.bincount(km.labels_))

# PCA (unsupervised)
pca = PCA(n_components=2).fit(X)
X_pca = pca.transform(X)
print(f'PCA explained variance: {pca.explained_variance_ratio_.sum():.2%}')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))
axes[0].scatter(X_pca[:,0], X_pca[:,1], c=km.labels_, cmap='viridis', alpha=0.7)
axes[0].set_title('K-Means Clusters (unsupervised)')
axes[1].scatter(X_pca[:,0], X_pca[:,1], c=y_true, cmap='viridis', alpha=0.7)
axes[1].set_title('True Labels (for comparison only)')
plt.tight_layout(); plt.show()
```

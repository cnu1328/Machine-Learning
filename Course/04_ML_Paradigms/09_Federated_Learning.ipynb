## 1. Intuition

**Problem:** Hospital/clinic/farm data cannot leave premises (privacy, regulations).

**Federated learning:** Train a global model by aggregating **local model updates** without sharing raw data.

1. Server sends global model to clients
2. Each client trains on local data
3. Clients send weight updates (not data) to server
4. Server averages updates → new global model

## 2. FedAvg Algorithm

$$w_{global} \leftarrow \sum_{k=1}^{K} \frac{n_k}{n} w_k$$

Weighted average of client model weights by dataset size.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

REPO = Path('../../').resolve()
plt.rcParams['figure.figsize'] = (8, 5)
rng = np.random.default_rng(42)
```

```python
# Simulate FedAvg with 5 clients
K = 5
n_samples = [100, 200, 150, 80, 120]
total = sum(n_samples)

# Global model weights (1D for demo)
w_global = np.zeros(10)

for round_num in range(10):
    client_weights = []
    for k in range(K):
        # Each client trains locally (simulated as global + noise)
        w_k = w_global + rng.normal(0, 0.1, 10)
        client_weights.append(w_k)
    # FedAvg
    w_global = sum(n_k / total * w_k for n_k, w_k in zip(n_samples, client_weights))

print(f'Global model after 10 rounds: {np.round(w_global, 3)}')
```

## 1. Intuition

**Meta-learning:** Train a model that can **quickly adapt** to new tasks with few examples.

- **Transfer learning:** Source task → target task (one-time transfer)
- **Meta-learning:** Many tasks → learn initialization that adapts fast to any new task

**MAML (Model-Agnostic Meta-Learning):**
Find $\theta^*$ such that one gradient step on any new task gives good performance:
$$\theta^* = \arg\min_\theta \sum_{\mathcal{T}_i} L_{\mathcal{T}_i}(\theta - \alpha \nabla_\theta L_{\mathcal{T}_i}(\theta))$$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

REPO = Path('../../').resolve()
plt.rcParams['figure.figsize'] = (8, 5)
rng = np.random.default_rng(42)
```

```python
# MAML intuition demo: find initialization good for multiple tasks
# Each task: y = a*x + b with different a, b
tasks = [(rng.uniform(1,3), rng.uniform(-2,2)) for _ in range(5)]

# Bad initialization
w_bad = np.array([0.0, 0.0])
# Good meta-initialization (average of task-optimal weights)
w_meta = np.array([np.mean([a for a,_ in tasks]), np.mean([b for _,b in tasks])])

for label, w in [('Bad init', w_bad), ('Meta init', w_meta)]:
    total_loss = 0
    for a, b in tasks:
        x = rng.uniform(-1, 1, 20)
        y = a * x + b + rng.normal(0, 0.1, 20)
        pred = w[0] * x + w[1]
        total_loss += np.mean((pred - y)**2)
    print(f'{label}: avg loss across tasks = {total_loss/len(tasks):.4f}')
```

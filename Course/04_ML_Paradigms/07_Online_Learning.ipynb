## 1. Intuition

**Online learning:** Update model as each new sample arrives — don't retrain from scratch.

**Use cases:**
- Streaming sensor data
- Fraud detection (patterns change daily)
- Recommendation systems (new users/items constantly)

**Concept drift:** Data distribution changes over time. Model must adapt.

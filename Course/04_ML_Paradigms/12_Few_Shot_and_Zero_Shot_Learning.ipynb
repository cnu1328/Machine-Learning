## 1. Definitions

**Few-shot:** Classify new categories from **K examples per class** (e.g., 5-way 1-shot = 5 classes, 1 example each).

**Zero-shot:** Classify **without any examples** — using text descriptions or attributes.

**CLIP (Contrastive Language-Image Pre-training):** Maps images and text to same embedding space. Zero-shot: compare image to text prompts like "a satellite image of aquaculture pond" vs "forest".

## 2. Prototypical Networks

For N-way K-shot:
1. Compute **prototype** (mean embedding) for each class from K support examples
2. Classify query by nearest prototype

$$c_k = \frac{1}{K}\sum_{i} f_\theta(x_i^k) \quad \text{(prototype for class k)}$$

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

REPO = Path('../../').resolve()
plt.rcParams['figure.figsize'] = (8, 5)
rng = np.random.default_rng(42)
````

````python
# Prototypical network demo
def compute_prototypes(X_support, y_support, n_way=3):
    prototypes = {}
    for c in range(n_way):
        prototypes[c] = X_support[y_support == c].mean(axis=0)
    return prototypes

def classify_query(x_query, prototypes):
    dists = {c: np.linalg.norm(x_query - p) for c, p in prototypes.items()}
    return min(dists, key=dists.get)

from sklearn.datasets import load_iris
X, y = load_iris(return_X_y=True)

# 3-way 3-shot
prototypes = compute_prototypes(X[:9], y[:9], n_way=3)
correct = sum(classify_query(X[i], prototypes) == y[i] for i in range(9, 30))
print(f'Prototypical network accuracy (3-way 3-shot): {correct/21:.4f}')
````

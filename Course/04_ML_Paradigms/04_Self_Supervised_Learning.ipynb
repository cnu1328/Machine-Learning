## 1. Intuition

**Self-supervised learning** creates labels **from the data itself** — no human annotation.

Examples of pretext tasks:
- Predict rotation angle of image (0°, 90°, 180°, 270°)
- Predict masked patch in image (BERT for images — MAE)
- Contrastive: augmented views of same image should be similar

After pretraining, fine-tune on small labeled dataset.

## 2. Contrastive Learning (SimCLR)

For image $x$, create two augmentations $x_i, x_j$:

- **Positive pair:** $(x_i, x_j)$ from same image
- **Negative pairs:** $(x_i, x_k)$ from different images

**Loss (InfoNCE):** pull positives together, push negatives apart

$$L = -\log \frac{\exp(\text{sim}(z_i, z_j)/\tau)}{\sum_k \exp(\text{sim}(z_i, z_k)/\tau)}$$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

REPO = Path('../../').resolve()
plt.rcParams['figure.figsize'] = (8, 5)
rng = np.random.default_rng(42)
```

```python
# Simplified contrastive loss demo
def contrastive_loss(z_i, z_j, z_negatives, temperature=0.5):
    """InfoNCE for one positive pair vs negatives."""
    sim_pos = np.dot(z_i, z_j) / (np.linalg.norm(z_i) * np.linalg.norm(z_j))
    sim_neg = [np.dot(z_i, z_n) / (np.linalg.norm(z_i) * np.linalg.norm(z_n)) for z_n in z_negatives]
    logits = np.array([sim_pos] + sim_neg) / temperature
    logits = logits - logits.max()  # stability
    exp_logits = np.exp(logits)
    return -np.log(exp_logits[0] / exp_logits.sum())

z_anchor = rng.normal(0, 1, 64)
z_positive = z_anchor + rng.normal(0, 0.1, 64)  # similar
z_negs = [rng.normal(0, 1, 64) for _ in range(5)]  # different
print(f'Contrastive loss: {contrastive_loss(z_anchor, z_positive, z_negs):.4f}')
```

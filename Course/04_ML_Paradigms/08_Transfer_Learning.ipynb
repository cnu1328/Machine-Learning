## 1. Intuition

**Transfer learning:** Use knowledge from **source task** to improve **target task**.

- Pretrain on ImageNet (1.2M images, 1000 classes)
- Fine-tune on your aquaculture pond dataset (500 images)

**Why it works:** Early layers learn universal features (edges, textures). Later layers are task-specific.

## 2. Strategies

| Strategy | Description | When |
|----------|-------------|------|
| **Feature extraction** | Freeze pretrained layers, train new head | Small target dataset |
| **Fine-tuning** | Unfreeze some/all layers, train with low LR | Medium target dataset |
| **Full training** | Train from scratch | Large target dataset or very different domain |

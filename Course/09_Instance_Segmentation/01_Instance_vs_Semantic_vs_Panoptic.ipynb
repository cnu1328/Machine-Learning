## Three Segmentation Paradigms

| Type | Output | Separates instances? | Example |
|------|--------|---------------------|--------|
| **Semantic** | Class per pixel | No | All water = class 1 |
| **Instance** | Mask per object + ID | Yes | Pond 1, Pond 2, Pond 3 |
| **Panoptic** | Semantic (stuff) + instance (things) | Yes for things | Road + building₁ + building₂ |

**Stuff vs Things:**
- **Stuff:** Amorphous regions (sky, road, water) — semantic only
- **Things:** Countable objects (car, pond, building) — need instance IDs

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import torch
import torch.nn as nn
import torch.nn.functional as F

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
rng = np.random.default_rng(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
# Simulate adjacent ponds: semantic merges, instance separates
H, W = 64, 128
semantic = torch.zeros(H, W, dtype=torch.long)
semantic[:, 20:45] = 1  # one water blob (two ponds merged)

inst1 = torch.zeros(H, W, dtype=torch.bool)
inst1[:, 20:32] = True
inst2 = torch.zeros(H, W, dtype=torch.bool)
inst2[:, 33:45] = True

fig, axes = plt.subplots(1, 3, figsize=(12, 3))
axes[0].imshow(semantic, cmap='Blues'); axes[0].set_title('Semantic (merged)'); axes[0].axis('off')
axes[1].imshow(inst1.numpy(), cmap='Blues'); axes[1].set_title('Instance 1'); axes[1].axis('off')
axes[2].imshow(inst2.numpy(), cmap='Oranges'); axes[2].set_title('Instance 2'); axes[2].axis('off')
plt.tight_layout(); plt.show()
```

## Your water-bodies-detection vs Instance Seg

**Instance seg (Mask R-CNN):** Predict separate mask per pond instance.

**Your approach (Module 07):** Dual-head semantic — aqua interior + bund boundary — separates adjacent ponds without instance IDs.

| Approach | Pros | Cons |
|----------|------|------|
| Dual-head boundary | Simpler labels, fast inference | No instance ID for tracking |
| Mask R-CNN | Unique ID per pond | Needs instance annotations |
| Panoptic | Full scene understanding | Heavy compute, complex labels |

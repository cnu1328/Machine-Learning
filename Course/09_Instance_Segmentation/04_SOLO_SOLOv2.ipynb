## SOLO (Wang et al., 2020)

**Segmenting Objects by Locations** — no bounding box proposals.

**Idea:** Divide feature map into $S \times S$ grid. Each cell responsible for objects whose **center** falls in that cell.

**Per grid cell predict:**
- Category label
- Instance mask ($E.g.$ 256 channels → upsample to full mask)

**SOLOv2 improvements:**
- Dynamic conv kernels per instance
- Better multi-scale handling
- Faster and higher AP than SOLO v1

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import torch
import torch.nn as nn
import torch.nn.functional as F

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
rng = np.random.default_rng(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
S = 8
# Each grid cell: class + mask coefficients (concept)
grid_classes = torch.randint(0, 5, (S, S))
grid_mask_dim = 32
grid_masks = torch.randn(S, S, grid_mask_dim)

fig, ax = plt.subplots(figsize=(4, 4))
ax.imshow(grid_classes.numpy(), cmap='tab10'); ax.set_title('SOLO grid: class per cell')
plt.show()
```

## SOLO vs Mask R-CNN

| | Mask R-CNN | SOLOv2 |
|---|-----------|--------|
| Proposals | RPN required | None |
| Parallelism | Sequential RoI | Fully parallel grid |
| Speed | Slower | Faster |
| Small objects | Good with FPN | Improving |

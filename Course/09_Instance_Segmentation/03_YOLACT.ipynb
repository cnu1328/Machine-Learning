## YOLACT (Bolya et al., 2019)

**Real-time instance segmentation** by factorizing masks:

$$M_i = \sigma\left(\sum_k c_{i,k} \cdot P_k\right)$$

- **Prototype masks** $P_k$: $K$ shared basis masks (e.g. 32 prototypes, H×W)
- **Coefficients** $c_i$: per-instance vector from detection branch
- **Assembly:** Linear combination → instance mask

**Benefits:** Single forward pass like YOLO; much faster than Mask R-CNN.

**Trade-off:** Slightly lower mAP on COCO vs Mask R-CNN.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import torch
import torch.nn as nn
import torch.nn.functional as F

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
rng = np.random.default_rng(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
# YOLACT mask assembly (concept)
K, H, W = 4, 32, 32
prototypes = torch.randn(K, H, W)
coeffs = torch.tensor([0.8, -0.3, 0.5, 0.1])  # instance 1
mask = torch.sigmoid((coeffs[:, None, None] * prototypes).sum(0))
plt.imshow(mask.numpy(), cmap='Blues'); plt.title('Assembled instance mask'); plt.colorbar(); plt.show()
```

## When to Use YOLACT

- Video / real-time aquaculture monitoring
- Edge deployment where Mask R-CNN too slow
- Acceptable if mask boundaries less precise than UNet++

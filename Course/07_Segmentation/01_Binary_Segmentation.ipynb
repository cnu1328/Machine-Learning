## 1. Binary Segmentation

**Task:** Classify each pixel as foreground (1) or background (0).

**Output:** Single channel mask of shape $(H, W)$ with values in $\{0, 1\}$ or probabilities $[0, 1]$.

**Examples:** Water body detection, building footprint, road extraction, medical tumor segmentation.

**Loss:** Binary cross-entropy + Dice (your water-bodies project).

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
rng = np.random.default_rng(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
# Synthetic binary segmentation data
N, H, W = 32, 64, 64
images = torch.randn(N, 3, H, W)
masks = (torch.sin(torch.linspace(0, 3.14, W)) > 0).float()
masks = masks.unsqueeze(0).unsqueeze(0).expand(N, 1, H, W)

fig, axes = plt.subplots(1, 3, figsize=(10, 3))
axes[0].imshow(images[0].permute(1,2,0).numpy().clip(0,1)); axes[0].set_title('Image'); axes[0].axis('off')
axes[1].imshow(masks[0,0], cmap='Blues'); axes[1].set_title('Binary Mask'); axes[1].axis('off')
axes[2].imshow((images[0].permute(1,2,0).numpy()*0.5 + masks[0,0].numpy()[:,:,None]*0.5).clip(0,1))
axes[2].set_title('Overlay'); axes[2].axis('off')
plt.tight_layout(); plt.show()
```

## Boundary Detection

Predict a thin band at object edges. Used to:
- Separate touching instances (your aquaculture ponds)
- Improve edge quality in medical imaging
- Post-process semantic masks

**Your project:** `masks_boundary/` — dilated polygon edges at configurable width (meters per pixel).

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
rng = np.random.default_rng(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
# Simulate aqua mask + boundary mask
aqua = torch.zeros(1, 1, 64, 64)
aqua[0, 0, 15:50, 15:50] = 1.0
# Boundary = morphological gradient (dilate - erode approximation)
pad = aqua.clone()
pad[:,:,1:,:] = torch.maximum(pad[:,:,1:,:], aqua[:,:,:-1,:])
pad[:,:,:-1,:] = torch.maximum(pad[:,:,:-1,:], aqua[:,:,1:,:])
boundary = (pad - aqua).clamp(0, 1)

fig, axes = plt.subplots(1, 3, figsize=(10, 3))
for ax, m, t in zip(axes, [aqua[0,0], boundary[0,0], aqua[0,0]+boundary[0,0]*0.5], ['Aqua','Boundary','Combined']):
    ax.imshow(m.numpy(), cmap='Blues'); ax.set_title(t); ax.axis('off')
plt.tight_layout(); plt.show()
```

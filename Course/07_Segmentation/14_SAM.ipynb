## SAM (Kirillov et al., 2023)

**Foundation model** for segmentation:
- **Prompts:** points, boxes, or coarse masks
- **Zero-shot:** works on unseen categories without fine-tuning

**GeoSpatial use:** Prompt with point on pond → get mask. Useful for rapid labeling and inference on new regions.

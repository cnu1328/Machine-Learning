## Three Segmentation Types

**Semantic:** Classify every pixel (all ponds = same class "water").

**Instance:** Separate individual objects (pond 1, pond 2, pond 3).

**Panoptic:** Semantic + instance (stuff + things).

|------|---------------------|--------|
| Semantic | No | 07 (this module) |
| Instance | Yes | 09 |
| Panoptic | Both | 09 |

## Learning Objectives

1. Understand independent labels per pixel
2. Use sigmoid (not softmax) per channel
3. Distinguish from multi-class

## Multi-Label vs Multi-Class

| | Multi-Class | Multi-Label |
|---|-------------|-------------|
| Labels per pixel | Exactly 1 | 0 or more |
| Output activation | Softmax | Sigmoid |
| Loss | Cross-entropy | BCE per channel |
| Example | Land cover class | Cloud AND shadow |

**Your water-bodies:** Two binary heads (aqua + boundary) = multi-label style with 2 channels.

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
rng = np.random.default_rng(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
# Multi-label: pixel can have aqua=1 AND boundary=1 (bund pixel)
target = torch.zeros(4, 2, 32, 32)
target[:, 0, 10:22, 10:22] = 1.0  # aqua interior
target[:, 1, 9:23, 9:23] = 1.0   # boundary ring

fig, axes = plt.subplots(1, 2, figsize=(8, 3))
axes[0].imshow(target[0,0], cmap='Blues'); axes[0].set_title('Aqua channel'); axes[0].axis('off')
axes[1].imshow(target[0,1], cmap='Reds'); axes[1].set_title('Boundary channel'); axes[1].axis('off')
plt.tight_layout(); plt.show()
```

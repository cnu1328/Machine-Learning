## FCN (Long et al., 2015)

**Key idea:** Replace fully connected layers of classification CNN with 1×1 convolutions → dense prediction map.

**Upsampling:** Transposed convolution (deconv) or bilinear upsampling to restore spatial resolution.

**FCN-8s:** Combine predictions from pool3, pool4, pool5 (skip connections).

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
rng = np.random.default_rng(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
class SimpleFCN(nn.Module):
    def __init__(self, n_classes=2):
        super().__init__()
        self.enc = nn.Sequential(
            nn.Conv2d(3, 32, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
            nn.Conv2d(32, 64, 3, padding=1), nn.ReLU(), nn.MaxPool2d(2),
        )
        self.head = nn.Conv2d(64, n_classes, 1)
        self.up = nn.ConvTranspose2d(n_classes, n_classes, 4, stride=2, padding=1)
    def forward(self, x):
        x = self.enc(x)
        x = self.head(x)
        x = self.up(self.up(x))
        return x

model = SimpleFCN(2)
x = torch.randn(2, 3, 64, 64)
print(f'FCN output: {model(x).shape}')  # (2, 2, 64, 64)
```

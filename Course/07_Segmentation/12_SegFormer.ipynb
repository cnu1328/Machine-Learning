## SegFormer (Xie et al., 2021)

**Encoder:** Hierarchical Vision Transformer (MiT)
**Decoder:** Lightweight MLP (no complex upsampling)

Efficient, strong on ADE20K/Cityscapes. Bridge to Module 10 Transformers.

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader, TensorDataset

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
rng = np.random.default_rng(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
print('SegFormer architecture:')
print('  Encoder: Mix Transformer (MiT-B0 to MiT-B5)')
print('  Decoder: MLP on each scale → fuse → classify')
print('  No positional encoding interpolation needed')
print('\nFor GeoSpatial: growing use for land cover mapping at scale')
```

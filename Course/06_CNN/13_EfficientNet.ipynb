## EfficientNet (Tan & Le, 2019)

**Compound scaling:** uniformly scale depth ($\alpha$), width ($\beta$), and resolution ($\gamma$):

$$\text{depth} = \alpha^\phi, \quad \text{width} = \beta^\phi, \quad \text{resolution} = \gamma^\phi$$

Subject to $\alpha \beta^2 \gamma^2 \approx 2$ and baseline FLOPS constraint.

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
import torchvision
import torchvision.transforms as T

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
from torchvision.models import efficientnet_b0, EfficientNet_B0_Weights

model = efficientnet_b0(weights=EfficientNet_B0_Weights.IMAGENET1K_V1)
print(f'EfficientNet-B0 params: {sum(p.numel() for p in model.parameters()):,}')

# Compound scaling variants
for name in ['efficientnet_b0', 'efficientnet_b1', 'efficientnet_b2']:
    m = getattr(__import__('torchvision.models', fromlist=[name]), name)(weights=None)
    print(f'{name}: {sum(p.numel() for p in m.parameters()):,} params')
```

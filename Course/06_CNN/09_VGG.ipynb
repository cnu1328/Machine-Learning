## VGG (Simonyan & Zisserman, 2014)

**Key idea:** Only 3×3 conv layers, stacked deeply. Smaller filters + more layers = more non-linearity with fewer parameters than large filters.

VGG-16: 13 conv + 3 FC layers. ~138M parameters.

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
import torchvision
import torchvision.transforms as T

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
def vgg_block(in_ch, out_ch, n_convs):
    layers = []
    for i in range(n_convs):
        layers += [nn.Conv2d(in_ch if i==0 else out_ch, out_ch, 3, padding=1), nn.ReLU()]
    layers.append(nn.MaxPool2d(2))
    return nn.Sequential(*layers)

class VGGSmall(nn.Module):
    def __init__(self, num_classes=10):
        super().__init__()
        self.features = nn.Sequential(
            vgg_block(3, 64, 2), vgg_block(64, 128, 2), vgg_block(128, 256, 2),
        )
        self.classifier = nn.Sequential(nn.Flatten(), nn.Linear(256*4*4, 256), nn.ReLU(), nn.Linear(256, num_classes))
    def forward(self, x): return self.classifier(self.features(x))

model = VGGSmall()
print(f'VGG-Small params: {sum(p.numel() for p in model.parameters()):,}')
```

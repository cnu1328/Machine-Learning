## 1. Digital Images as Tensors

**Grayscale:** shape $(H, W)$ — height × width pixels, values 0–255 or 0.0–1.0

**RGB:** shape $(H, W, 3)$ or $(3, H, W)$ — 3 color channels

**Batch:** $(N, C, H, W)$ — N images, C channels, H×W spatial

**PyTorch convention:** $(N, C, H, W)$ — channels first

**Your water-bodies project:** $(N, 6, 512, 512)$ — 6 Planet spectral bands

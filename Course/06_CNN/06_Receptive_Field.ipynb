## Receptive Field (RF)

The RF of a neuron = region of input image that influences its value.

**Recursive formula:**
$$RF_l = RF_{l-1} + (K_l - 1) \times \prod_{i=1}^{l-1} S_i$$

Deep networks see larger portions of the input — critical for context in segmentation.

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

plt.rcParams['figure.figsize'] = (8, 5)
rng = np.random.default_rng(42)
```

```python
def compute_rf(layers):
    """layers: list of (kernel_size, stride) per layer"""
    rf, jump = 1, 1
    for K, S in layers:
        rf = rf + (K - 1) * jump
        jump = jump * S
    return rf

# AlexNet-like stack
layers = [(11,4),(3,1),(3,1),(3,1),(3,1),(3,1)]
print(f'Receptive field after {len(layers)} layers: {compute_rf(layers)} pixels')

# water-bodies SE-ResNet50: RF covers entire 512x512 tile after encoder
print('SE-ResNet50 encoder RF >> 512 → each output pixel sees full tile context')
```

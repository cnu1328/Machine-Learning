## What Filters Learn

- **Layer 1:** edges, colors, simple textures
- **Layer 2-3:** corners, patterns
- **Deeper layers:** object parts, complex textures

**Feature map:** output of one filter applied to input — highlights where that pattern appears.

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
import torchvision
import torchvision.transforms as T

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
from torchvision import models

# Load pretrained AlexNet, visualize first conv filters
model = models.alexnet(weights=models.AlexNet_Weights.IMAGENET1K_V1)
first_conv = model.features[0].weight.data  # (64, 3, 11, 11)
print(f'First conv weights shape: {first_conv.shape}')

fig, axes = plt.subplots(4, 8, figsize=(12, 6))
for i, ax in enumerate(axes.flat):
    if i < first_conv.shape[0]:
        f = first_conv[i].permute(1, 2, 0)
        f = (f - f.min()) / (f.max() - f.min() + 1e-8)
        ax.imshow(f.numpy()); ax.axis('off')
plt.suptitle('AlexNet First Layer Filters (learned edge detectors)'); plt.tight_layout(); plt.show()
```

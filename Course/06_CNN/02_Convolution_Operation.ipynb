## 1. Mathematical Definition

2D convolution (cross-correlation in DL):

$$(I * K)_{ij} = \sum_m \sum_n I_{i+m, j+n} \cdot K_{m,n}$$

- $I$ = input image
- $K$ = kernel/filter (learnable weights)
- Output = feature map

**Key insight:** Convolution detects local patterns (edges, textures) by sliding a small kernel across the image.

```python
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path

plt.rcParams['figure.figsize'] = (8, 5)
rng = np.random.default_rng(42)
```

```python
def conv2d_manual(image, kernel):
    """Naive 2D convolution (valid, no padding, stride=1)."""
    h, w = image.shape
    kh, kw = kernel.shape
    out_h, out_w = h - kh + 1, w - kw + 1
    output = np.zeros((out_h, out_w))
    for i in range(out_h):
        for j in range(out_w):
            output[i, j] = np.sum(image[i:i+kh, j:j+kw] * kernel)
    return output

# Edge detection kernel (Sobel X)
image = np.random.rand(8, 8)
sobel_x = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
result = conv2d_manual(image, sobel_x)
print(f'Input: {image.shape}, Kernel: {sobel_x.shape}, Output: {result.shape}')
```

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data import DataLoader
import torchvision
import torchvision.transforms as T

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
# PyTorch equivalent
img = torch.randn(1, 1, 8, 8)  # (N, C, H, W)
kernel = torch.tensor(sobel_x, dtype=torch.float32).unsqueeze(0).unsqueeze(0)
out = F.conv2d(img, kernel)
print(f'PyTorch conv2d output shape: {out.shape}')
```

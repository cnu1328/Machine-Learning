## Metrics

**Precision:** $P = \frac{TP}{TP+FP}$ — of predicted positives, how many correct?

**Recall:** $R = \frac{TP}{TP+FN}$ — of actual positives, how many found?

**F1:** $F1 = 2\frac{PR}{P+R}$

**Accuracy misleading** when classes imbalanced.

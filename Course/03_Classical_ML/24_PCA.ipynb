## Derivation

1. Center data: $X_c = X - \bar{X}$
2. Covariance: $C = \frac{1}{n}X_c^T X_c$
3. Eigendecompose: $C = V\Lambda V^T$
4. Project: $X_{PCA} = X_c V_k$

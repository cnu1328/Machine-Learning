## Idea

Train network to reconstruct normal data. Anomalies have high reconstruction error.

(Full autoencoder implementation in Module 05 Deep Learning.)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

REPO = Path('../../').resolve()
plt.rcParams['figure.figsize'] = (8, 5)
rng = np.random.default_rng(42)
```

```python
# PCA as linear autoencoder baseline
from sklearn.decomposition import PCA

X_n = rng.normal(0, 1, (500, 10))
X_n[0] = 10  # inject anomaly

pca = PCA(n_components=5).fit(X_n[1:])
reconstructed = pca.inverse_transform(pca.transform(X_n))
errors = ((X_n - reconstructed)**2).mean(axis=1)
print('Reconstruction error (normal):', errors[1:6])
print('Reconstruction error (anomaly):', errors[0])
```

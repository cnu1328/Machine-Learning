## 1. Algorithm

1. Store all training samples
2. For new point, find $k$ nearest neighbors (by distance)
3. **Classification:** majority vote
4. **Regression:** average of neighbor targets

No training phase — lazy learning.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

REPO = Path('../../').resolve()
plt.rcParams['figure.figsize'] = (8, 5)
rng = np.random.default_rng(42)
```

```python
def knn_predict(X_train, y_train, X_test, k=5):
    preds = []
    for x in X_test:
        dists = np.sqrt(((X_train - x)**2).sum(axis=1))
        nearest = np.argsort(dists)[:k]
        preds.append(np.round(np.mean(y_train[nearest])))
    return np.array(preds)

from sklearn.datasets import load_iris
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)

pred = knn_predict(X_train, y_train, X_test, k=5)
print(f'KNN scratch accuracy: {(pred == y_test).mean():.4f}')
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5).fit(X_train, y_train)
print(f'sklearn KNN: {knn.score(X_test, y_test):.4f}')
```

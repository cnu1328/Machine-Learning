## 1. Why Polynomial Regression

Linear model in **expanded features**, not raw inputs:
$$\hat{y} = w_0 + w_1 x + w_2 x^2 + w_3 x^3 + \cdots$$

Captures curvature without changing the linear regression solver.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

REPO = Path('../../').resolve()
plt.rcParams['figure.figsize'] = (8, 5)
rng = np.random.default_rng(42)
```

```python
salary = pd.read_csv(REPO / 'salary.csv')
print(salary.head())

from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline

X = salary[['Level']].values
y = salary['Salary'].values

for deg in [1, 2, 3, 10]:
    pipe = Pipeline([('poly', PolynomialFeatures(deg)), ('lr', LinearRegression())])
    pipe.fit(X, y)
    print(f'Degree {deg:2d}: R² = {pipe.score(X, y):.4f}')
```

```python
fig, ax = plt.subplots()
ax.scatter(X, y, label='Data')
X_line = np.linspace(X.min(), X.max(), 200).reshape(-1, 1)
for deg, c in [(1,'blue'),(3,'green'),(10,'red')]:
    pipe = Pipeline([('poly', PolynomialFeatures(deg)), ('lr', LinearRegression())])
    pipe.fit(X, y); ax.plot(X_line, pipe.predict(X_line), c=c, label=f'deg={deg}')
ax.legend(); ax.set_title('Polynomial Regression — Overfitting at degree 10'); plt.show()
```

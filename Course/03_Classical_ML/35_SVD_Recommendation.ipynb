## Matrix Factorization

Rating matrix $R \approx U \Sigma V^T$

- $U$ = user factors
- $V$ = item factors
- Low-rank approximation captures latent preferences

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

REPO = Path('../../').resolve()
plt.rcParams['figure.figsize'] = (8, 5)
rng = np.random.default_rng(42)
```

```python
# Simple user-item matrix
R = np.array([[5, 3, 0, 1], [4, 0, 0, 1], [1, 1, 0, 5], [0, 0, 5, 4], [0, 1, 5, 4]], dtype=float)

U, S, Vt = np.linalg.svd(R, full_matrices=False)

k = 2  # latent factors
R_approx = U[:, :k] @ np.diag(S[:k]) @ Vt[:k, :]
print('Original:\n', R)
print('\nApprox (k=2):\n', np.round(R_approx, 1))
```

## 1. Objective

$$\min_{C} \sum_{k=1}^{K}\sum_{x_i \in C_k} \|x_i - \mu_k\|^2$$

Alternate: assign points to nearest centroid, update centroids as cluster means.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

REPO = Path('../../').resolve()
plt.rcParams['figure.figsize'] = (8, 5)
rng = np.random.default_rng(42)
```

```python
def kmeans(X, k, max_iters=100):
    centroids = X[rng.choice(len(X), k, replace=False)]
    for _ in range(max_iters):
        dists = np.array([[np.linalg.norm(x - c) for c in centroids] for x in X])
        labels = dists.argmin(axis=1)
        new_centroids = np.array([X[labels==i].mean(axis=0) if (labels==i).any() else centroids[i] for i in range(k)])
        if np.allclose(centroids, new_centroids): break
        centroids = new_centroids
    return labels, centroids

from sklearn.datasets import load_iris
X = load_iris().data
labels, _ = kmeans(X, 3)
print('Cluster sizes:', np.bincount(labels))
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

for k in range(2, 8):
    km = KMeans(n_clusters=k, random_state=42, n_init=10).fit(X)
    print(f'k={k}: silhouette={silhouette_score(X, km.labels_):.3f}')
```

## Metrics

**R²:** $R^2 = 1 - \frac{SS_{res}}{SS_{tot}} = 1 - \frac{\sum(y_i - \hat{y}_i)^2}{\sum(y_i - \bar{y})^2}$

**Adjusted R²:** $R^2_{adj} = 1 - (1-R^2)\frac{n-1}{n-p-1}$ (penalizes extra features)

**RMSE:** $\sqrt{\frac{1}{n}\sum(y_i - \hat{y}_i)^2}$

**MAE:** $\frac{1}{n}\sum|y_i - \hat{y}_i|$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

REPO = Path('../../').resolve()
plt.rcParams['figure.figsize'] = (8, 5)
rng = np.random.default_rng(42)
```

```python
from sklearn.ensemble import IsolationForest

X_normal = rng.normal(0, 1, (200, 2))
X_outliers = rng.uniform(-4, 4, (20, 2))
X_all = np.vstack([X_normal, X_outliers])

iso = IsolationForest(contamination=0.1, random_state=42).fit(X_all)
labels = iso.predict(X_all)

plt.scatter(X_all[:,0], X_all[:,1], c=labels, cmap='coolwarm', alpha=0.7)
plt.title('Isolation Forest — Anomaly Detection'); plt.show()
```

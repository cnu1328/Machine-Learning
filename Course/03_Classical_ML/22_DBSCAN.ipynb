## Learning Objectives

1. Understand density-based clustering
2. Handle arbitrary cluster shapes
3. Identify outliers

## Algorithm

- **eps:** neighborhood radius
- **min_samples:** min points to form dense region
- Core points → clusters; border points → assigned; rest → noise

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

REPO = Path('../../').resolve()
plt.rcParams['figure.figsize'] = (8, 5)
rng = np.random.default_rng(42)
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_moons

X, _ = make_moons(n_samples=300, noise=0.05, random_state=42)
db = DBSCAN(eps=0.2, min_samples=5).fit(X)

plt.scatter(X[:,0], X[:,1], c=db.labels_, cmap='viridis', s=20)
plt.title(f'DBSCAN — {len(set(db.labels_))-1} clusters'); plt.show()
```

DBSCAN finds arbitrary shapes and labels outliers. Sensitive to eps.

## Detection Paradigm Comparison

| Model | Type | Speed | mAP | Best For |
|-------|------|-------|-----|----------|
| Faster R-CNN | Two-stage | Medium | High | Sparse, precise boxes |
| SSD | Single-stage | Fast | Medium | Embedded systems |
| RetinaNet | Single-stage | Medium | High | Multi-scale objects |
| YOLOv8 | Single-stage | Very fast | High | Production GeoSpatial |
| DETR | Transformer | Medium | High | Research, no NMS |
| RT-DETR | Transformer | Fast | High | Real-time transformer |

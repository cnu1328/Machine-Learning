## YOLO v1 (Redmon et al., 2016)

Divide image into $S \times S$ grid. Each cell predicts:
- $B$ bounding boxes (x, y, w, h, confidence)
- Class probabilities

**Loss:** MSE on coords + classification.

**YOLO v2:** Anchor boxes, batch norm, multi-scale training.

**YOLO v3:** Feature pyramid, 3 scales, logistic class prediction.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import torch
import torch.nn as nn
import torch.nn.functional as F

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
rng = np.random.default_rng(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
S, B, C = 7, 2, 3  # grid, boxes per cell, classes
# Prediction tensor per image: S x S x (B*5 + C)
pred = torch.randn(S, S, B * 5 + C)

# Decode one cell prediction (concept)
cell_pred = pred[3, 3]
conf = torch.sigmoid(cell_pred[4])  # first box confidence
cls = torch.softmax(cell_pred[B*5:], dim=0)
print(f'Cell (3,3) conf={conf:.3f}, best class={cls.argmax().item()}')
```

## YOLO Philosophy

Treat detection as **regression** — one network, one pass, real-time.

Limitation of v1: struggles with small objects and nearby duplicates (one object per cell).

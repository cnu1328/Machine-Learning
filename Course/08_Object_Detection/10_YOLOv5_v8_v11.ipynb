## GeoSpatial: Building Detection

**Workflow:**
1. Annotate buildings in QGIS/Label Studio → YOLO txt format
2. Train YOLOv8 on 512×512 satellite tiles
3. Run sliding-window inference on large GeoTIFF
4. Export to ONNX for deployment

**vs water-bodies segmentation:** Detection gives bounding boxes for quick counts; segmentation gives exact boundaries.

## Production Tips

- Tile large images with overlap
- Use lower conf threshold at inference, NMS to filter
- Monitor mAP@0.5 on validation tiles
- Augment: rotation, flip, color jitter (careful with NIR bands)

## SSD (Liu et al., 2016)

**Single shot:** Predict class + box at multiple feature map scales in one forward pass.

**Default boxes:** Fixed anchor sizes per layer (e.g. 38×38 layer for small objects, 10×10 for large).

**Multi-scale:** Early layers → small objects; deep layers → large objects.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import torch
import torch.nn as nn
import torch.nn.functional as F

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
rng = np.random.default_rng(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
# SSD-style multi-scale heads (concept)
scales = [(38, 38), (19, 19), (10, 10), (5, 5), (3, 3), (1, 1)]
anchors_per_loc = 4
for h, w in scales:
    n_anchors = h * w * anchors_per_loc
    print(f'Feature map {h}x{w}: {n_anchors:,} default boxes')
```

## Trade-offs

| | Faster R-CNN | SSD |
|---|-------------|-----|
| Stages | 2 | 1 |
| Speed | Slower | Faster |
| Small objects | Better (FPN helps) | Harder without FPN |

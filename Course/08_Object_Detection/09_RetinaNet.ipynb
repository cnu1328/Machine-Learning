## RetinaNet (Lin et al., 2017)

**Problem:** Extreme foreground/background imbalance (~100k:1 anchors) — CE loss dominated by easy negatives.

**Focal Loss:** $(1-p_t)^\gamma$ down-weights easy examples.

**FPN:** Top-down pathway + lateral connections → rich multi-scale features.

**Result:** First single-stage detector to match two-stage mAP on COCO.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import torch
import torch.nn as nn
import torch.nn.functional as F

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
rng = np.random.default_rng(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
def focal_loss_binary(logits, targets, gamma=2.0, alpha=0.25):
    bce = F.binary_cross_entropy_with_logits(logits, targets, reduction='none')
    p = torch.sigmoid(logits)
    p_t = p * targets + (1 - p) * (1 - targets)
    return (alpha * (1 - p_t) ** gamma * bce).mean()

# Simulated anchor labels: 1000 bg, 10 fg
logits = torch.randn(1010)
targets = torch.cat([torch.zeros(1000), torch.ones(10)])
print(f'Focal loss (imbalanced): {focal_loss_binary(logits, targets):.4f}')
```

## R-CNN (Girshick et al., 2014)

**Pipeline:**
1. **Region proposals** (~2000) via Selective Search
2. **Warp** each region to fixed size (227×227)
3. **CNN** (AlexNet) → feature vector
4. **SVM** classifier per class
5. **BBox regressor** per class

**Problems:** Slow (CNN run per region), multi-stage training, no end-to-end gradient.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import torch
import torch.nn as nn
import torch.nn.functional as F

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
rng = np.random.default_rng(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
# R-CNN concept: many proposals, each classified separately
n_proposals = 2000
proposal_time = 2.0  # selective search seconds
forward_per_region = 0.05  # AlexNet forward seconds
total = proposal_time + n_proposals * forward_per_region
print(f'R-CNN inference ~{total:.0f}s per image (conceptual)')
print('Bottleneck: no shared computation across regions')
```

## Historical Significance

First to show CNN features + region proposals beat hand-crafted features on PASCAL VOC.

**Your takeaway:** Detection = **where to look** (proposals) + **what is it** (classifier).

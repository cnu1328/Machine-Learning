## DETR (Carion et al., 2020)

**Paradigm shift:** No anchors, no NMS.

**Architecture:**
1. CNN backbone → feature map
2. Transformer encoder
3. Transformer decoder with **N learned object queries** (e.g. 100)
4. Each query → class + box (cxcywh normalized)

**Training:** Hungarian algorithm matches predictions to GT (bipartite matching loss).

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import torch
import torch.nn as nn
import torch.nn.functional as F

plt.rcParams['figure.figsize'] = (8, 5)
torch.manual_seed(42)
rng = np.random.default_rng(42)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f'Device: {device}')
```

```python
# DETR output format (concept)
N_queries = 100
num_classes = 91  # COCO
logits = torch.randn(N_queries, num_classes)
boxes = torch.sigmoid(torch.randn(N_queries, 4))  # cxcywh in [0,1]

# Inference: take queries where class != 'no-object'
probs = logits.softmax(-1)
scores, labels = probs[:, :-1].max(-1)
keep = scores > 0.7
print(f'High-confidence detections: {keep.sum().item()}')
```

## Pros & Cons

| Pros | Cons |
|------|------|

| Simple pipeline | Large objects easier than small |
| Set-based loss | Higher compute than YOLO |
